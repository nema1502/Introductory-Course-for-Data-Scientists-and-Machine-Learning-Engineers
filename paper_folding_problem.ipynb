### Problema 5 Guardar en la Lista
Hasta ahora solo hemos usado el último valor después de 43 difracciones, pero también necesitamos el valor del proceso para visualizarlo en el gráfico. Al código que usa la instrucción for, agregue el código que registra un total de 44 valores de proceso.

"Insinuación"

En primer lugar, cree una lista vacía.
El valor predoblado se añade entonces a la lista.
Los valores del proceso de doblar el papel hasta n veces en una sentencia for se añaden a la lista.
Por último, es importante comprobar que hay 44 valores almacenados en la lista mediante la función len.

```python
THICKNESS = 0.00008  # Grosor inicial en metros
folds = 43           # Número de dobleces

# Inicializar lista con el grosor inicial
thickness_list = [THICKNESS]

# Calcular grosor en cada doblez usando un bucle for
for _ in range(folds):
    THICKNESS *= 2  # Grosor se duplica en cada iteración
    thickness_list.append(THICKNESS)  # Añadir a la lista

# Verificar longitud de la lista (44 elementos: 0 a 43 dobleces)
print("Número de elementos en la lista:", len(thickness_list))
```
