```python
# ============================================================================
# CONFIGURACI√ìN Y DESCARGA CORRECTA DEL DATASET DE KAGGLE
# ============================================================================

def install_kaggle_api():
    """Instala la API oficial de Kaggle"""
    try:
        import kaggle
        print("‚úÖ Kaggle API ya disponible")
        return True
    except ImportError:
        print("üì¶ Instalando Kaggle API oficial...")
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "kaggle", "--quiet"])
            print("‚úÖ Kaggle API instalada exitosamente")
            return True
        except Exception as e:
            print(f"‚ùå Error instalando API: {e}")
            return False

def show_setup_instructions():
    """Muestra instrucciones de configuraci√≥n de Kaggle"""
    print("\nüìö CONFIGURACI√ìN DE KAGGLE API:")
    print("="*50)
    print("""
üîπ PASOS PARA CONFIGURAR KAGGLE:

1Ô∏è‚É£ CREAR TOKEN:
   ‚Ä¢ Ir a kaggle.com ‚Üí Account ‚Üí Create New API Token
   ‚Ä¢ Se descarga kaggle.json

2Ô∏è‚É£ CONFIGURAR CREDENCIALES:
   ‚Ä¢ Linux/Mac: mv kaggle.json ~/.kaggle/ && chmod 600 ~/.kaggle/kaggle.json
   ‚Ä¢ Windows: Mover a C:\\Users\\<usuario>\\.kaggle\\kaggle.json
   ‚Ä¢ Colab: Files ‚Üí Upload kaggle.json ‚Üí ejecutar setup

3Ô∏è‚É£ VERIFICAR:
   ‚Ä¢ Terminal: kaggle competitions list
   ‚Ä¢ Python: import kaggle; kaggle.api.authenticate()
""")

def try_kaggle_download():
    """Intenta descargar usando la API de Kaggle"""
    try:
        import kaggle
        api = kaggle.KaggleApi()
        api.authenticate()
        
        competition = 'house-prices-advanced-regression-techniques'
        print(f"üì• Descargando: {competition}")
        
        api.competition_download_files(competition, path='.', unzip=True)
        print("‚úÖ Descarga completada con API")
        return True
        
    except Exception as e:
        print(f"‚ùå Error con API de Kaggle: {e}")
        return False

def create_synthetic_dataset():
    """Crea dataset sint√©tico con estructura realista del dataset de Kaggle"""
    print("üìä Creando dataset sint√©tico con estructura realista...")
    
    np.random.seed(42)
    n_samples = 1460
    
    # Variables principales basadas en el dataset real
    overall_qual = np.random.choice(range(1, 11), n_samples, 
                                   p=[0.02, 0.03, 0.05, 0.08, 0.12, 0.15, 0.18, 0.14, 0.13, 0.10])
    
    year_built = np.random.choice(range(1872, 2011), n_samples)
    
    # √Åreas con distribuciones realistas
    gr_liv_area = np.random.lognormal(7.2, 0.4, n_samples)
    gr_liv_area = np.clip(gr_liv_area, 334, 5642)
    
    total_bsmt_sf = np.random.gamma(2.5, 400, n_samples)
    total_bsmt_sf = np.clip(total_bsmt_sf, 0, 6110)
    
    garage_area = np.random.gamma(2, 250, n_samples)
    garage_area = np.clip(garage_area, 0, 1418)
    
    first_flr_sf = np.random.gamma(3, 350, n_samples)
    first_flr_sf = np.clip(first_flr_sf, 334, 4692)
    
    # LotFrontage con 17.7% valores faltantes (como en dataset real)
    lot_frontage = np.random.gamma(4, 17.5, n_samples)
    lot_frontage = np.clip(lot_frontage, 21, 313)
    missing_mask = np.random.random(n_samples) < 0.177
    lot_frontage[missing_mask] = np.nan
    
    # MasVnrArea con algunos valores faltantes
    mas_vnr_area = np.where(np.random.random(n_samples) > 0.55,
                           np.random.exponential(103, n_samples), 0)
    mas_vnr_area = np.clip(mas_vnr_area, 0, 1600)
    missing_mask = np.random.random(n_samples) < 0.005
    mas_vnr_area[missing_mask] = np.nan
    
    # Calcular precio realista
    base_price = 80000
    price = (base_price +
             overall_qual * 22000 +      # Factor m√°s importante
             gr_liv_area * 70 +          # √Årea habitable
             total_bsmt_sf * 30 +        # S√≥tano
             garage_area * 50 +          # Garaje  
             (year_built - 1900) * 800 + # Edad
             np.random.normal(0, 30000, n_samples))  # Variabilidad
    
    price = np.clip(price, 34900, 755000)
    
    # Crear DataFrame con estructura similar al real
    data = pd.DataFrame({
        'Id': range(1, n_samples + 1),
        'MSSubClass': np.random.choice([20, 30, 40, 50, 60, 70, 80, 90, 120, 160, 180, 190], n_samples),
        'LotFrontage': lot_frontage,
        'LotArea': np.random.gamma(2, 5000, n_samples),
        'Street': np.random.choice(['Grvl', 'Pave'], n_samples, p=[0.006, 0.994]),
        'Alley': np.where(np.random.random(n_samples) > 0.93, 
                         np.random.choice(['Grvl', 'Pave'], n_samples), np.nan),
        'LotShape': np.random.choice(['Reg', 'IR1', 'IR2', 'IR3'], n_samples, p=[0.64, 0.30, 0.05, 0.01]),
        'Utilities': np.random.choice(['AllPub', 'NoSewr'], n_samples, p=[0.999, 0.001]),
        'LotConfig': np.random.choice(['Inside', 'Corner', 'CulDSac', 'FR2', 'FR3'], 
                                     n_samples, p=[0.74, 0.16, 0.06, 0.03, 0.01]),
        'Neighborhood': np.random.choice(['NAmes', 'CollgCr', 'OldTown', 'Edwards', 'Somerst', 
                                        'Gilbert', 'NridgHt', 'Sawyer', 'NWAmes', 'SawyerW',
                                        'BrkSide', 'Crawfor', 'Mitchel', 'NoRidge', 'Timber'], n_samples),
        'Condition1': np.random.choice(['Normal', 'Feedr', 'Artery', 'RRNn', 'PosN'], 
                                      n_samples, p=[0.86, 0.06, 0.04, 0.02, 0.02]),
        'BldgType': np.random.choice(['1Fam', '2fmCon', 'Duplex', 'TwnhsE', 'Twnhs'], 
                                    n_samples, p=[0.83, 0.02, 0.04, 0.08, 0.03]),
        'HouseStyle': np.random.choice(['1Story', '2Story', '1.5Fin', 'SLvl', 'SFoyer'], 
                                      n_samples, p=[0.50, 0.35, 0.10, 0.04, 0.01]),
        'OverallQual': overall_qual,
        'OverallCond': np.random.choice(range(1, 10), n_samples),
        'YearBuilt': year_built,
        'YearRemodAdd': np.minimum(year_built + np.random.randint(0, 40, n_samples), 2010),
        'RoofStyle': np.random.choice(['Gable', 'Hip', 'Gambrel', 'Mansard', 'Flat', 'Shed'], 
                                     n_samples, p=[0.79, 0.19, 0.007, 0.005, 0.004, 0.004]),
        'ExterQual': np.random.choice(['Ex', 'Gd', 'TA', 'Fa'], n_samples, p=[0.35, 0.30, 0.34, 0.01]),
        'ExterCond': np.random.choice(['Ex', 'Gd', 'TA', 'Fa', 'Po'], n_samples, p=[0.03, 0.30, 0.65, 0.019, 0.001]),
        'Foundation': np.random.choice(['PConc', 'CBlock', 'BrkTil', 'Wood', 'Slab'], 
                                      n_samples, p=[0.45, 0.41, 0.10, 0.02, 0.02]),
        'BsmtQual': np.random.choice(['Ex', 'Gd', 'TA', 'Fa'], n_samples, p=[0.08, 0.41, 0.43, 0.08]),
        'BsmtCond': np.random.choice(['Ex', 'Gd', 'TA', 'Fa', 'Po'], n_samples, p=[0.01, 0.05, 0.92, 0.019, 0.001]),
        'TotalBsmtSF': total_bsmt_sf,
        '1stFlrSF': first_flr_sf,
        '2ndFlrSF': np.where(np.random.random(n_samples) > 0.54, 
                           np.random.gamma(2, 350, n_samples), 0),
        'GrLivArea': gr_liv_area,
        'BsmtFullBath': np.random.choice([0, 1, 2], n_samples, p=[0.60, 0.35, 0.05]),
        'BsmtHalfBath': np.random.choice([0, 1, 2], n_samples, p=[0.91, 0.088, 0.002]),
        'FullBath': np.random.choice([0, 1, 2, 3, 4], n_samples, p=[0.03, 0.47, 0.44, 0.05, 0.01]),
        'HalfBath': np.random.choice([0, 1, 2], n_samples, p=[0.39, 0.58, 0.03]),
        'BedroomAbvGr': np.random.choice([0, 1, 2, 3, 4, 5], n_samples, p=[0.01, 0.03, 0.19, 0.44, 0.28, 0.05]),
        'KitchenAbvGr': np.random.choice([1, 2, 3], n_samples, p=[0.97, 0.025, 0.005]),
        'TotRmsAbvGrd': np.random.randint(3, 15, n_samples),
        'Functional': np.random.choice(['Typ', 'Min1', 'Min2', 'Mod'], n_samples, p=[0.93, 0.023, 0.021, 0.026]),
        'Fireplaces': np.random.choice([0, 1, 2, 3], n_samples, p=[0.47, 0.39, 0.12, 0.02]),
        'GarageType': np.random.choice(['Attchd', 'Detchd', 'BuiltIn', 'CarPort'], 
                                      n_samples, p=[0.60, 0.24, 0.06, 0.10]),
        'GarageYrBlt': np.where(garage_area > 0, 
                               np.minimum(year_built + np.random.randint(-5, 5, n_samples), 2010), np.nan),
        'GarageCars': np.random.choice([0, 1, 2, 3, 4], n_samples, p=[0.05, 0.17, 0.60, 0.17, 0.01]),
        'GarageArea': garage_area,
        'GarageQual': np.random.choice(['Ex', 'Gd', 'TA', 'Fa', 'Po'], n_samples, p=[0.01, 0.02, 0.95, 0.019, 0.001]),
        'GarageCond': np.random.choice(['Ex', 'Gd', 'TA', 'Fa', 'Po'], n_samples, p=[0.01, 0.02, 0.95, 0.019, 0.001]),
        'PavedDrive': np.random.choice(['Y', 'N', 'P'], n_samples, p=[0.91, 0.06, 0.03]),
        'WoodDeckSF': np.random.exponential(50, n_samples),
        'OpenPorchSF': np.random.exponential(46, n_samples),
        'EnclosedPorch': np.random.exponential(20, n_samples),
        '3SsnPorch': np.random.exponential(5, n_samples),
        'ScreenPorch': np.random.exponential(15, n_samples),
        'PoolArea': np.where(np.random.random(n_samples) > 0.99, 
                           np.random.exponential(100, n_samples), 0),
        'MiscVal': np.random.exponential(50, n_samples),
        'MasVnrArea': mas_vnr_area,
        'MoSold': np.random.choice(range(1, 13), n_samples),
        'YrSold': np.random.choice([2006, 2007, 2008, 2009, 2010], n_samples),
        'SaleType': np.random.choice(['WD', 'New', 'COD', 'ConLD'], n_samples, p=[0.83, 0.08, 0.05, 0.04]),
        'SaleCondition': np.random.choice(['Normal', 'Partial', 'Abnorml', 'Family'], 
                                         n_samples, p=[0.82, 0.085, 0.063, 0.032]),
        'SalePrice': price
    })
    
    # Redondear valores apropiados
    numeric_cols = ['LotArea', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 
                   'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch',
                   '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice']
    
    for col in numeric_cols:
        if col in data.columns:
            data[col] = data[col].round()
    
    print("‚úÖ Dataset sint√©tico creado con estructura completa del real")
    return data

# Ejecutar proceso de carga
print("üîÑ INICIANDO PROCESO DE CARGA DEL DATASET...")
print("="*50)

# Paso 1: Instalar API si es necesario
api_installed = install_kaggle_api()

# Paso 2: Mostrar instrucciones
show_setup_instructions()

# Paso 3: Intentar descarga autom√°tica
dataset_loaded = False
source = "No cargado"

if api_installed:
    print("\nüîÑ Intentando descarga con API de Kaggle...")
    success = try_kaggle_download()
    
    if success:
        try:
            data = pd.read_csv('train.csv')
            dataset_loaded = True
            source = "Kaggle Real (API)"
            print("‚úÖ Dataset real cargado desde API")
        except:
            print("‚ö†Ô∏è API ejecut√≥ pero train.csv no encontrado")

# Paso 4: Intentar carga desde archivo local
if not dataset_loaded:
    try:
        data = pd.read_csv('train.csv')
        dataset_loaded = True
        source = "Archivo Local"
        print("‚úÖ Dataset real cargado desde archivo local")
    except FileNotFoundError:
        print("üìù train.csv no encontrado localmente")

# Paso 5: Usar dataset sint√©tico
if not dataset_loaded:
    print("\nüîÑ Usando dataset sint√©tico para continuar an√°lisis...")
    data = create_synthetic_dataset()
    dataset_loaded = True
    source = "Dataset Sint√©tico"

# Verificar carga exitosa
if dataset_loaded:
    print(f"\nüìä DATASET CARGADO ({source}):")
    print(f"   ‚Ä¢ Forma: {data.shape}")
    print(f"   ‚Ä¢ Precio promedio: ${data['SalePrice'].mean():,.0f}")
    print(f"   ‚Ä¢ Variables num√©ricas: {len(data.select_dtypes(include=[np.number]).columns)}")
    print(f"   ‚Ä¢ Variables categ√≥ricas: {len(data.select_dtypes(include=[object]).columns)}")
    
    print("\n‚úÖ DATASET LISTO PARA AN√ÅLISIS")
else:
    print("‚ùå ERROR: No se pudo cargar ning√∫n dataset")
```

```python
# ============================================================================
# PROBLEMA 5: INVESTIGACI√ìN DE T√âRMINOS ESTAD√çSTICOS
# ============================================================================

print("\n" + "="*80)
print("‚úÖ PROBLEMA 5: INVESTIGACI√ìN DE T√âRMINOS ESTAD√çSTICOS")
print("="*80)

print("üìö DEFINICIONES ESTAD√çSTICAS FUNDAMENTALES:")
print("="*50)

print("""
üîπ CURTOSIS (Kurtosis):
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üìñ DEFINICI√ìN:
   Mide la concentraci√≥n de valores alrededor de la media y el "peso" 
   de las colas de la distribuci√≥n.

üìä INTERPRETACI√ìN:
   ‚Ä¢ Curtosis = 3: MESOC√öRTICA (distribuci√≥n normal)
   ‚Ä¢ Curtosis > 3: LEPTOC√öRTICA (m√°s puntiaguda, colas pesadas)
   ‚Ä¢ Curtosis < 3: PLATOC√öRTICA (m√°s plana, colas ligeras)

üßÆ C√ÅLCULO:
   Curtosis = E[(X-Œº)‚Å¥] / œÉ‚Å¥
   
üí° IMPLICACIONES:
   ‚Ä¢ Alta curtosis ‚Üí M√°s valores extremos (outliers)
   ‚Ä¢ Baja curtosis ‚Üí Distribuci√≥n m√°s uniforme
   ‚Ä¢ Importante para detecci√≥n de anomal√≠as

‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üîπ ASIMETR√çA (Skewness):
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üìñ DEFINICI√ìN:
   Mide la simetr√≠a de la distribuci√≥n respecto a la media.

üìä INTERPRETACI√ìN:
   ‚Ä¢ Asimetr√≠a = 0: SIM√âTRICA (distribuci√≥n normal)
   ‚Ä¢ Asimetr√≠a > 0: ASIM√âTRICA POSITIVA (cola derecha larga)
   ‚Ä¢ Asimetr√≠a < 0: ASIM√âTRICA NEGATIVA (cola izquierda larga)

üìè CLASIFICACI√ìN:
   ‚Ä¢ |Asimetr√≠a| < 0.5: Aproximadamente sim√©trica
   ‚Ä¢ 0.5 ‚â§ |Asimetr√≠a| < 1: Moderadamente asim√©trica  
   ‚Ä¢ |Asimetr√≠a| ‚â• 1: Altamente asim√©trica

üßÆ C√ÅLCULO:
   Asimetr√≠a = E[(X-Œº)¬≥] / œÉ¬≥

üí° IMPLICACIONES:
   ‚Ä¢ Positiva ‚Üí Media > Mediana (datos financieros t√≠picos)
   ‚Ä¢ Negativa ‚Üí Media < Mediana (escalas de satisfacci√≥n)
   ‚Ä¢ Afecta elecci√≥n de modelos estad√≠sticos

‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üîπ DISTRIBUCI√ìN NORMAL (Gaussiana):
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üìñ DEFINICI√ìN:
   Distribuci√≥n de probabilidad continua en forma de campana,
   completamente definida por media (Œº) y desviaci√≥n est√°ndar (œÉ).

üîî CARACTER√çSTICAS:
   ‚Ä¢ Forma de campana sim√©trica
   ‚Ä¢ Media = Mediana = Moda
   ‚Ä¢ Funci√≥n de densidad: f(x) = (1/œÉ‚àö2œÄ)e^(-(x-Œº)¬≤/2œÉ¬≤)

üìè REGLA 68-95-99.7:
   ‚Ä¢ 68% de datos dentro de Œº ¬± 1œÉ
   ‚Ä¢ 95% de datos dentro de Œº ¬± 2œÉ  
   ‚Ä¢ 99.7% de datos dentro de Œº ¬± 3œÉ

üéØ IMPORTANCIA:
   ‚Ä¢ Base del Teorema Central del L√≠mite
   ‚Ä¢ Supuesto en muchos m√©todos estad√≠sticos
   ‚Ä¢ Referencia para comparar otras distribuciones

üí° EN MACHINE LEARNING:
   ‚Ä¢ Regresi√≥n lineal asume errores normales
   ‚Ä¢ Muchos algoritmos funcionan mejor con datos normales
   ‚Ä¢ Facilita interpretaci√≥n de resultados

‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üîπ TRANSFORMACI√ìN LOGAR√çTMICA:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

üìñ PROP√ìSITO:
   Convertir distribuciones asim√©tricas en aproximadamente normales.

üîß MEC√ÅNICA:
   ‚Ä¢ Comprime valores grandes, expande valores peque√±os
   ‚Ä¢ Y_transformado = log(Y_original)
   ‚Ä¢ Especialmente √∫til para Y > 0

‚úÖ BENEFICIOS:
   ‚Ä¢ Reduce asimetr√≠a positiva
   ‚Ä¢ Estabiliza varianza
   ‚Ä¢ Mejora linealidad en relaciones
   ‚Ä¢ Facilita interpretaci√≥n porcentual

üéØ APLICACIONES COMUNES:
   ‚Ä¢ Precios de viviendas
   ‚Ä¢ Salarios e ingresos
   ‚Ä¢ Datos de poblaci√≥n
   ‚Ä¢ Concentraciones qu√≠micas

‚ö†Ô∏è CONSIDERACIONES:
   ‚Ä¢ Solo para valores positivos
   ‚Ä¢ Interpretaci√≥n cambia (efectos multiplicativos)
   ‚Ä¢ Puede requerir transformaci√≥n inversa para predicciones
""")

print("‚úÖ Problema 5: COMPLETADO")
```

```python
# ============================================================================
# PROBLEMA 6: VERIFICACI√ìN DE LA DISTRIBUCI√ìN
# ============================================================================

print("\n" + "="*80)
print("‚úÖ PROBLEMA 6: VERIFICACI√ìN DE LA DISTRIBUCI√ìN")
print("="*80)

# Usar dataset limpio para an√°lisis
sale_price = data_cleaned['SalePrice']

print("üìä 1. AN√ÅLISIS DESCRIPTIVO DE SALEPRICE:")
print("-" * 42)
print(f"   ‚Ä¢ Observaciones: {len(sale_price):,}")
print(f"   ‚Ä¢ Media: ${sale_price.mean():,.0f}")
print(f"   ‚Ä¢ Mediana: ${sale_price.median():,.0f}")
print(f"   ‚Ä¢ Desviaci√≥n est√°ndar: ${sale_price.std():,.0f}")
print(f"   ‚Ä¢ M√≠nimo: ${sale_price.min():,.0f}")
print(f"   ‚Ä¢ M√°ximo: ${sale_price.max():,.0f}")
print(f"   ‚Ä¢ Rango: ${sale_price.max() - sale_price.min():,.0f}")

# Calcular medidas de forma
original_kurtosis = sale_price.kurtosis()
original_skewness = sale_price.skew()

print(f"\nüìê 2. MEDIDAS DE FORMA - DISTRIBUCI√ìN ORIGINAL:")
print("-" * 48)
print(f"   ‚Ä¢ Curtosis: {original_kurtosis:.4f}")
print(f"   ‚Ä¢ Asimetr√≠a: {original_skewness:.4f}")

# Interpretaciones autom√°ticas
def interpret_skewness(skew_value):
    if skew_value > 1:
        return "FUERTEMENTE ASIM√âTRICA POSITIVA", "üî¥"
    elif skew_value > 0.5:
        return "MODERADAMENTE ASIM√âTRICA POSITIVA", "üü†"
    elif skew_value > 0:
        return "LIGERAMENTE ASIM√âTRICA POSITIVA", "üü°"
    elif skew_value > -0.5:
        return "APROXIMADAMENTE SIM√âTRICA", "üü¢"
    elif skew_value > -1:
        return "MODERADAMENTE ASIM√âTRICA NEGATIVA", "üü†"
    else:
        return "FUERTEMENTE ASIM√âTRICA NEGATIVA", "üî¥"

def interpret_kurtosis(kurt_value):
    if kurt_value > 3:
        excess = kurt_value - 3
        if excess > 2:
            return f"LEPTOC√öRTICA FUERTE (exceso: {excess:.2f})", "üî¥"
        else:
            return f"LEPTOC√öRTICA MODERADA (exceso: {excess:.2f})", "üü†"
    elif kurt_value < 3:
        deficit = 3 - kurt_value
        if deficit > 1:
            return f"PLATOC√öRTICA FUERTE (d√©ficit: {deficit:.2f})", "üî¥"
        else:
            return f"PLATOC√öRTICA MODERADA (d√©ficit: {deficit:.2f})", "üü†"
    else:
        return "MESOC√öRTICA (similar a normal)", "üü¢"

skew_interp, skew_icon = interpret_skewness(original_skewness)
kurt_interp, kurt_icon = interpret_kurtosis(original_kurtosis)

print(f"   ‚Ä¢ Interpretaci√≥n asimetr√≠a: {skew_icon} {skew_interp}")
print(f"   ‚Ä¢ Interpretaci√≥n curtosis: {kurt_icon} {kurt_interp}")

# Visualizaci√≥n de la distribuci√≥n original
print(f"\nüìä 3. VISUALIZACI√ìN DE DISTRIBUCI√ìN ORIGINAL:")
print("-" * 45)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Histograma con estad√≠sticas
ax1.hist(sale_price, bins=50, density=True, alpha=0.7, color='skyblue', 
         edgecolor='black', label='Histograma')
ax1.axvline(sale_price.mean(), color='red', linestyle='--', linewidth=2, 
           label=f'Media: ${sale_price.mean():,.0f}')
ax1.axvline(sale_price.median(), color='green', linestyle='--', linewidth=2, 
           label=f'Mediana: ${sale_price.median():,.0f}')

# Agregar curva normal para comparaci√≥n
x_norm = np.linspace(sale_price.min(), sale_price.max(), 100)
normal_curve = stats.norm.pdf(x_norm, sale_price.mean(), sale_price.std())
ax1.plot(x_norm, normal_curve, 'orange', linewidth=2, label='Normal te√≥rica')

ax1.set_xlabel('Precio de Venta ($)')
ax1.set_ylabel('Densidad')
ax1.set_title(f'Distribuci√≥n Original de SalePrice\nAsimetr√≠a: {original_skewness:.3f}, Curtosis: {original_kurtosis:.3f}', 
              fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Box plot
box_plot = ax2.boxplot(sale_price, patch_artist=True)
box_plot['boxes'][0].set_facecolor('lightcoral')
ax2.set_ylabel('Precio de Venta ($)')
ax2.set_title('Box Plot - Distribuci√≥n Original', fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Q-Q plot contra distribuci√≥n normal
stats.probplot(sale_price, dist="norm", plot=ax3)
ax3.set_title('Q-Q Plot: Original vs Distribuci√≥n Normal', fontweight='bold')
ax3.grid(True, alpha=0.3)

# Transformaci√≥n logar√≠tmica
print(f"\nüîÑ 4. APLICANDO TRANSFORMACI√ìN LOGAR√çTMICA:")
print("-" * 45)

log_sale_price = np.log(sale_price)
log_kurtosis = log_sale_price.kurtosis()
log_skewness = log_sale_price.skew()

print(f"   ‚Ä¢ Curtosis log-transformada: {log_kurtosis:.4f}")
print(f"   ‚Ä¢ Asimetr√≠a log-transformada: {log_skewness:.4f}")

# Interpretaciones de la transformaci√≥n
log_skew_interp, log_skew_icon = interpret_skewness(log_skewness)
log_kurt_interp, log_kurt_icon = interpret_kurtosis(log_kurtosis)

print(f"   ‚Ä¢ Nueva interpretaci√≥n asimetr√≠a: {log_skew_icon} {log_skew_interp}")
print(f"   ‚Ä¢ Nueva interpretaci√≥n curtosis: {log_kurt_icon} {log_kurt_interp}")

# Histograma log-transformado
ax4.hist(log_sale_price, bins=50, density=True, alpha=0.7, color='lightgreen', 
         edgecolor='black', label='Log-transformado')
ax4.axvline(log_sale_price.mean(), color='red', linestyle='--', linewidth=2, 
           label=f'Media: {log_sale_price.mean():.3f}')
ax4.axvline(log_sale_price.median(), color='green', linestyle='--', linewidth=2, 
           label=f'Mediana: {log_sale_price.median():.3f}')

# Curva normal para log-transformado
x_log_norm = np.linspace(log_sale_price.min(), log_sale_price.max(), 100)
log_normal_curve = stats.norm.pdf(x_log_norm, log_sale_price.mean(), log_sale_price.std())
ax4.plot(x_log_norm, log_normal_curve, 'orange', linewidth=2, label='Normal te√≥rica')

ax4.set_xlabel('Log(Precio de Venta)')
ax4.set_ylabel('Densidad')
ax4.set_title(f'Distribuci√≥n Log-transformada\nAsimetr√≠a: {log_skewness:.3f}, Curtosis: {log_kurtosis:.3f}', 
              fontweight='bold')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Evaluaci√≥n cuantitativa de la mejora
print(f"\nüìà 5. EVALUACI√ìN DE LA TRANSFORMACI√ìN:")
print("-" * 40)

skew_improvement = abs(original_skewness) - abs(log_skewness)
kurt_improvement = abs(original_kurtosis - 3) - abs(log_kurtosis - 3)

print(f"   ‚Ä¢ Mejora en asimetr√≠a: {skew_improvement:+.4f}")
print(f"   ‚Ä¢ Mejora en curtosis: {kurt_improvement:+.4f}")
print(f"   ‚Ä¢ Asimetr√≠a se {'MEJOR√ì' if skew_improvement > 0 else 'EMPEOR√ì'}: {'‚úÖ' if skew_improvement > 0 else '‚ùå'}")
print(f"   ‚Ä¢ Curtosis se {'MEJOR√ì' if kurt_improvement > 0 else 'EMPEOR√ì'}: {'‚úÖ' if kurt_improvement > 0 else '‚ùå'}")

# Test de normalidad (opcional)
try:
    from scipy.stats import shapiro, jarque_bera
    
    print(f"\nüß™ 6. TESTS DE NORMALIDAD:")
    print("-" * 25)
    
    # Test de Jarque-Bera (mejor para muestras grandes)
    jb_orig_stat, jb_orig_p = jarque_bera(sale_price)
    jb_log_stat, jb_log_p = jarque_bera(log_sale_price)
    
    print(f"   ‚Ä¢ Jarque-Bera Original: estad√≠stico={jb_orig_stat:.2f}, p-valor={jb_orig_p:.2e}")
    print(f"   ‚Ä¢ Jarque-Bera Log-transformado: estad√≠stico={jb_log_stat:.2f}, p-valor={jb_log_p:.2e}")
    print(f"   ‚Ä¢ Mejora en normalidad: {'‚úÖ' if jb_log_p > jb_orig_p else '‚ùå'}")
    
except:
    print("   ‚Ä¢ Tests de normalidad no disponibles")

# Explicaci√≥n pr√°ctica
print(f"\nüí° 7. EXPLICACI√ìN PR√ÅCTICA CON T√âRMINOS DE DISTRIBUCI√ìN NORMAL:")
print("-" * 65)
print(f"""
üî∏ DISTRIBUCI√ìN ORIGINAL (T√≠pica de datos financieros):
   ‚Ä¢ Precios muestran patr√≥n com√∫n en mercados inmobiliarios
   ‚Ä¢ ASIMETR√çA POSITIVA: Mayor√≠a de casas en rango medio-bajo de precios,
     pero casas de lujo crean una 'cola larga' hacia valores altos
   ‚Ä¢ ALEJADA DE NORMALIDAD: No sigue el patr√≥n de campana de Gauss
   ‚Ä¢ Media > Mediana debido a valores extremos altos

üî∏ DESPU√âS DE TRANSFORMACI√ìN LOGAR√çTMICA:
   ‚Ä¢ Se acerca m√°s a la DISTRIBUCI√ìN NORMAL (campana de Gauss)
   ‚Ä¢ Reduce la ASIMETR√çA al 'comprimir' logar√≠tmicamente valores extremos
   ‚Ä¢ Estabiliza la VARIANZA a lo largo del rango de datos
   ‚Ä¢ Media y mediana se aproximan m√°s

üî∏ BENEFICIOS PARA MACHINE LEARNING:
   ‚Ä¢ MEJOR PERFORMANCE: Regresi√≥n lineal, redes neuronales funcionan mejor
   ‚Ä¢ REDUCE OUTLIERS: Menor influencia de casas extremadamente caras
   ‚Ä¢ MEJORA CONVERGENCIA: Algoritmos de optimizaci√≥n convergen m√°s r√°pido
   ‚Ä¢ INTERPRETACI√ìN: Cambios se interpretan como porcentajes vs. valores absolutos

üî∏ INTERPRETACI√ìN DE COEFICIENTES (Log-transformado):
   ‚Ä¢ Un aumento de 1 unidad en X ‚Üí cambio de Œ≤% en precio
   ‚Ä¢ Ejemplo: Œ≤=0.1 ‚Üí 10% de aumento en precio
   ‚Ä¢ M√°s intuitivo para stakeholders inmobiliarios

üî∏ CU√ÅNDO USAR TRANSFORMACI√ìN LOGAR√çTMICA:
   ‚Ä¢ Datos con asimetr√≠a positiva fuerte (como precios, salarios)
   ‚Ä¢ Relaciones multiplicativas en lugar de aditivas
   ‚Ä¢ Cuando se busca estabilizar varianza
   ‚Ä¢ Para mejorar interpretabilidad en t√©rminos porcentuales
""")

print("\n‚úÖ Problema 6: COMPLETADO")
```

```python
# ============================================================================
# PROBLEMA 7: AN√ÅLISIS DE CORRELACIONES
# ============================================================================

print("\n" + "="*80)
print("‚úÖ PROBLEMA 7: AN√ÅLISIS COMPLETO DE CORRELACIONES")
print("="*80)

# An√°lisis comprehensivo de correlaciones
numeric_data = data_cleaned.select_dtypes(include=[np.number])
corr_matrix = numeric_data.corr()

print(f"üîç 1. CONFIGURACI√ìN DEL AN√ÅLISIS:")
print("-" * 35)
print(f"   ‚Ä¢ Variables num√©ricas analizadas: {len(numeric_data.columns)}")
print(f"   ‚Ä¢ Observaciones: {len(numeric_data):,}")
print(f"   ‚Ä¢ Correlaciones totales: {(len(numeric_data.columns) * (len(numeric_data.columns) - 1)) // 2}")

# Matriz de correlaci√≥n completa
print(f"\nüìä 2. MATRIZ DE CORRELACI√ìN COMPLETA:")
print("-" * 40)

plt.figure(figsize=(20, 16))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)

# Crear heatmap con anotaciones
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='RdYlBu_r', center=0, 
            square=True, fmt='.2f', cbar_kws={'label': 'Coeficiente de Correlaci√≥n'},
            annot_kws={'size': 8}, linewidths=0.1)

plt.title('Matriz de Correlaci√≥n Completa (Tri√°ngulo Inferior)\nDataset de Precios de Viviendas - Ames, Iowa', 
          fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Top correlaciones con SalePrice
print(f"\nüèÜ 3. TOP 15 CORRELACIONES CON SALEPRICE:")
print("-" * 45)

saleprice_corr = corr_matrix['SalePrice'].abs().sort_values(ascending=False)
top15_features = saleprice_corr.head(16).index[1:]  # Excluir SalePrice mismo

print("Ranking  Caracter√≠stica               Correlaci√≥n    Fuerza        Direcci√≥n")
print("-" * 80)

for i, feature in enumerate(top15_features, 1):
    correlation = corr_matrix['SalePrice'][feature]
    
    # Clasificar fuerza de correlaci√≥n
    abs_corr = abs(correlation)
    if abs_corr >= 0.8:
        strength = "MUY FUERTE"
        strength_icon = "üî¥"
    elif abs_corr >= 0.6:
        strength = "FUERTE    "
        strength_icon = "üü†"
    elif abs_corr >= 0.4:
        strength = "MODERADA  "
        strength_icon = "üü°"
    elif abs_corr >= 0.2:
        strength = "D√âBIL     "
        strength_icon = "üîµ"
    else:
        strength = "MUY D√âBIL "
        strength_icon = "‚ö™"
    
    direction = "Positiva ‚ÜóÔ∏è" if correlation > 0 else "Negativa ‚ÜòÔ∏è"
    
    print(f"   {i:2d}    {feature:<25}    {correlation:8.4f}    {strength_icon} {strength}    {direction}")

# An√°lisis de correlaci√≥n - Top 10
top10_features = list(top15_features)[:10]
top10_with_price = ['SalePrice'] + top10_features
corr_top10 = data_cleaned[top10_with_price].corr()

print(f"\nüìä 4. MATRIZ DE CORRELACI√ìN - TOP 10 CARACTER√çSTICAS:")
print("-" * 55)

plt.figure(figsize=(14, 12))
sns.heatmap(corr_top10, annot=True, cmap='RdYlBu_r', center=0, 
            square=True, fmt='.3f', cbar_kws={'label': 'Correlaci√≥n'},
            annot_kws={'size': 12}, linewidths=0.2)
plt.title('Matriz de Correlaci√≥n - Top 10 Caracter√≠sticas vs SalePrice', 
          fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Explicaciones detalladas de caracter√≠sticas principales
print(f"\nüìã 5. AN√ÅLISIS DETALLADO DE CARACTER√çSTICAS PRINCIPALES:")
print("-" * 58)

# Diccionario de explicaciones basado en dataset real de Kaggle
feature_explanations = {
    'OverallQual': {
        'desc': 'Calidad general de materiales y acabados de la casa',
        'scale': 'Escala 1-10 (1=Muy Pobre, 10=Excelente)',
        'impact': 'Factor m√°s predictivo del precio'
    },
    'GrLivArea': {
        'desc': '√Årea de vivienda sobre el nivel del suelo',
        'scale': 'Pies cuadrados',
        'impact': 'Cada pie¬≤ adicional impacta directamente el precio'
    },
    'TotalBsmtSF': {
        'desc': '√Årea total del s√≥tano',
        'scale': 'Pies cuadrados',
        'impact': 'Espacio adicional utilizable, muy valorado'
    },
    '1stFlrSF': {
        'desc': '√Årea del primer piso',
        'scale': 'Pies cuadrados', 
        'impact': 'Correlacionado con GrLivArea, √°rea principal'
    },
    'GarageArea': {
        'desc': '√Årea del garaje',
        'scale': 'Pies cuadrados',
        'impact': 'Protecci√≥n y almacenamiento, comodidad valorada'
    },
    'YearBuilt': {
        'desc': 'A√±o de construcci√≥n original de la casa',
        'scale': 'A√±o (1872-2010)',
        'impact': 'Casas m√°s nuevas = precios m√°s altos'
    },
    'FullBath': {
        'desc': 'N√∫mero de ba√±os completos sobre el nivel del suelo',
        'scale': 'Cantidad (0-4+)',
        'impact': 'Comodidad esencial, impacto no lineal'
    },
    'TotRmsAbvGrd': {
        'desc': 'Total de habitaciones sobre el nivel del suelo',
        'scale': 'Cantidad (no incluye ba√±os)',
        'impact': 'Indicador de tama√±o y funcionalidad'
    },
    'Fireplaces': {
        'desc': 'N√∫mero de chimeneas en la propiedad',
        'scale': 'Cantidad (0-4+)',
        'impact': 'Amenidad de lujo, valor est√©tico'
    },
    'YearRemodAdd': {
        'desc': 'A√±o de remodelaci√≥n/adici√≥n m√°s reciente',
        'scale': 'A√±o (igual a YearBuilt si nunca se remodel√≤)',
        'impact': 'Modernizaci√≥n a√±ade valor significativo'
    },
    'GarageCars': {
        'desc': 'Capacidad del garaje en n√∫mero de autom√≥viles',
        'scale': 'Cantidad (0-5)',
        'impact': 'Fuertemente correlacionado con GarageArea'
    },
    'MasVnrArea': {
        'desc': '√Årea de revestimiento de mamposter√≠a',
        'scale': 'Pies cuadrados',
        'impact': 'Material de calidad, mejora est√©tica'
    },
    '2ndFlrSF': {
        'desc': '√Årea del segundo piso',
        'scale': 'Pies cuadrados (0 si no hay segundo piso)',
        'impact': '√Årea adicional, pero menos valorada que primer piso'
    },
    'OpenPorchSF': {
        'desc': '√Årea del porche abierto',
        'scale': 'Pies cuadrados',
        'impact': 'Espacio exterior, menor impacto en precio'
    },
    'HalfBath': {
        'desc': 'N√∫mero de medios ba√±os (solo inodoro y lavabo)',
        'scale': 'Cantidad (0-2)',
        'impact': 'Comodidad adicional, menor que FullBath'
    }
}

for i, feature in enumerate(top10_features, 1):
    correlation = corr_matrix['SalePrice'][feature]
    
    # Determinar importancia econ√≥mica
    abs_corr = abs(correlation)
    if abs_corr >= 0.7:
        importance = "CR√çTICA"
        importance_icon = "üî¥"
    elif abs_corr >= 0.5:
        importance = "ALTA"
        importance_icon = "üü†"
    elif abs_corr >= 0.3:
        importance = "MODERADA"
        importance_icon = "üü°"
    else:
        importance = "BAJA"
        importance_icon = "üîµ"
    
    feature_info = feature_explanations.get(feature, {
        'desc': 'Descripci√≥n no disponible',
        'scale': 'N/A',
        'impact': 'Impacto a determinar'
    })
    
    print(f"\n{i:2d}. {importance_icon} {feature} (r = {correlation:+.4f}) - Importancia: {importance}")
    print(f"    üìù {feature_info['desc']}")
    print(f"    üìè Escala: {feature_info['scale']}")
    print(f"    üí∞ Impacto: {feature_info['impact']}")
    
    # Estad√≠sticas adicionales para caracter√≠sticas clave
    if i <= 5:  # Top 5
        feature_data = data_cleaned[feature]
        print(f"    üìä Rango: {feature_data.min():.0f} - {feature_data.max():.0f}")
        print(f"    üìà Media: {feature_data.mean():.1f}, Mediana: {feature_data.median():.1f}")

# Intercorrelaciones altas (multicolinealidad)
print(f"\nüîó 6. AN√ÅLISIS DE INTERCORRELACIONES (MULTICOLINEALIDAD):")
print("-" * 58)

high_corr_pairs = []
for i in range(len(corr_top10.columns)):
    for j in range(i+1, len(corr_top10.columns)):
        feature1 = corr_top10.columns[i]
        feature2 = corr_top10.columns[j]
        correlation = corr_top10.iloc[i, j]
        
        if abs(correlation) > 0.7 and 'SalePrice' not in [feature1, feature2]:
            high_corr_pairs.append((feature1, feature2, correlation))

if high_corr_pairs:
    high_corr_sorted = sorted(high_corr_pairs, key=lambda x: abs(x[2]), reverse=True)
    
    print("Caracter√≠stica 1             Caracter√≠stica 2             Correlaci√≥n   Interpretaci√≥n")
    print("-" * 90)
    
    # Explicaciones l√≥gicas de correlaciones altas
    logical_explanations = {
        ('GrLivArea', '1stFlrSF'): "√Årea total incluye primer piso",
        ('TotalBsmtSF', '1stFlrSF'): "S√≥tanos grandes ‚Üí primeros pisos grandes",
        ('GarageArea', 'GarageCars'): "M√°s capacidad ‚Üí m√°s √°rea requerida",
        ('YearBuilt', 'YearRemodAdd'): "Sin remodelaci√≥n ‚Üí mismo a√±o",
        ('OverallQual', 'YearBuilt'): "Casas nuevas ‚Üí mejor calidad",
        ('FullBath', 'TotRmsAbvGrd'): "M√°s ba√±os ‚Üí m√°s habitaciones totales"
    }
    
    for i, (f1, f2, corr) in enumerate(high_corr_sorted[:7], 1):
        key = (f1, f2) if (f1, f2) in logical_explanations else (f2, f1)
        explanation = logical_explanations.get(key, "Relaci√≥n a investigar")
        
        print(f"{f1:<27} {f2:<27} {corr:8.4f}   {explanation}")
        
        if i <= 3:  # An√°lisis detallado para top 3
            print(f"   üí° Implicaci√≥n: Posible redundancia para modelos ML")
    
    print(f"\n‚ö†Ô∏è  ADVERTENCIA MULTICOLINEALIDAD:")
    print(f"   ‚Ä¢ {len(high_corr_pairs)} pares con correlaci√≥n > 0.7")
    print(f"   ‚Ä¢ Considerar t√©cnicas de regularizaci√≥n (Ridge/Lasso)")
    print(f"   ‚Ä¢ Posible eliminaci√≥n de variables redundantes")
    print(f"   ‚Ä¢ VIF (Variance Inflation Factor) recomendado para an√°lisis detallado")
    
else:
    print("‚úÖ No se encontraron intercorrelaciones problem√°ticas (>0.7)")

# Visualizaci√≥n de relaciones m√°s fuertes
print(f"\nüìä 7. DIAGRAMAS DE DISPERSI√ìN - RELACIONES M√ÅS FUERTES:")
print("-" * 58)

top4_features = top10_features[:4]
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, feature in enumerate(top4_features):
    correlation = corr_matrix['SalePrice'][feature]
    
    # Crear scatter plot
    axes[i].scatter(data_cleaned[feature], data_cleaned['SalePrice'], 
                   alpha=0.6, s=30, color='steelblue', edgecolors='black', linewidth=0.1)
    
    # L√≠nea de tendencia
    z = np.polyfit(data_cleaned[feature], data_cleaned['SalePrice'], 1)
    p = np.poly1d(z)
    axes[i].plot(data_cleaned[feature], p(data_cleaned[feature]), 
                "r--", alpha=0.8, linewidth=3, label=f'Tendencia (r¬≤={correlation**2:.3f})')
    
    axes[i].set_xlabel(feature, fontweight='bold')
    axes[i].set_ylabel('SalePrice ($)', fontweight='bold')
    axes[i].set_title(f'{feature} vs SalePrice\nCorrelaci√≥n: r = {correlation:.4f}', 
                     fontweight='bold', fontsize=12)
    axes[i].grid(True, alpha=0.3)
    axes[i].legend()
    
    # Formatear eje Y como moneda
    axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
    
    # Estad√≠sticas en el gr√°fico
    axes[i].text(0.05, 0.95, f'R¬≤ = {correlation**2:.3f}\nN = {len(data_cleaned):,}', 
                transform=axes[i].transAxes, fontsize=10, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.suptitle('Relaciones M√°s Fuertes con Precio de Venta', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# An√°lisis de correlaciones d√©biles o negativas
print(f"\nüìâ 8. CORRELACIONES D√âBILES O NEGATIVAS INTERESANTES:")
print("-" * 56)

weak_negative_corr = []
all_correlations = corr_matrix['SalePrice'].sort_values()

# Encontrar correlaciones negativas y muy d√©biles
negative_corr = all_correlations[all_correlations < 0]
weak_positive_corr = all_correlations[(all_correlations > 0) & (all_correlations < 0.1)]

print("üîª CORRELACIONES NEGATIVAS:")
for feature, corr_val in negative_corr.items():
    if feature != 'SalePrice':
        print(f"   ‚Ä¢ {feature:<25}: {corr_val:+.4f}")

print("\nüîµ CORRELACIONES POSITIVAS D√âBILES (<0.1):")
for feature, corr_val in weak_positive_corr.items():
    if feature != 'SalePrice':
        print(f"   ‚Ä¢ {feature:<25}: {corr_val:+.4f}")

# Resumen ejecutivo final
strongest_feature = top10_features[0]
strongest_corr = corr_matrix['SalePrice'][strongest_feature]

print(f"\nüéØ 9. RESUMEN EJECUTIVO - AN√ÅLISIS DE CORRELACIONES:")
print("="*60)

print(f"""
üìä HALLAZGOS PRINCIPALES:

ü•á PREDICTOR M√ÅS FUERTE:
   ‚Ä¢ {strongest_feature} (r = {strongest_corr:.4f})
   ‚Ä¢ Explica {strongest_corr**2:.1%} de la varianza en precio
   ‚Ä¢ Variable cr√≠tica para cualquier modelo predictivo

üìà DISTRIBUCI√ìN DE FUERZAS:
   ‚Ä¢ Correlaciones fuertes (|r| ‚â• 0.6): {len([f for f in top15_features if abs(corr_matrix['SalePrice'][f]) >= 0.6])} caracter√≠sticas
   ‚Ä¢ Correlaciones moderadas (0.4 ‚â§ |r| < 0.6): {len([f for f in top15_features if 0.4 <= abs(corr_matrix['SalePrice'][f]) < 0.6])} caracter√≠sticas
   ‚Ä¢ Correlaciones d√©biles (|r| < 0.4): {len([f for f in top15_features if abs(corr_matrix['SalePrice'][f]) < 0.4])} caracter√≠sticas

üè† CATEGOR√çAS M√ÅS INFLUYENTES:
   ‚Ä¢ CALIDAD Y CONDICI√ìN: OverallQual, ExterQual, BsmtQual
   ‚Ä¢ TAMA√ëOS Y √ÅREAS: GrLivArea, TotalBsmtSF, GarageArea, 1stFlrSF
   ‚Ä¢ TEMPORALES: YearBuilt, YearRemodAdd (modernizaci√≥n)
   ‚Ä¢ COMODIDADES: FullBath, Fireplaces, TotRmsAbvGrd

‚ö†Ô∏è  CONSIDERACIONES T√âCNICAS:
   ‚Ä¢ Multicolinealidad detectada en {len(high_corr_pairs)} pares
   ‚Ä¢ Regularizaci√≥n recomendada (Ridge/Lasso/Elastic Net)
   ‚Ä¢ Feature selection podr√≠a mejorar interpretabilidad
   ‚Ä¢ Cross-validation esencial para validaci√≥n

üéØ RECOMENDACIONES PARA MODELOS:
   ‚Ä¢ Usar top 10-15 caracter√≠sticas para modelo inicial
   ‚Ä¢ Considerar transformaci√≥n logar√≠tmica de SalePrice
   ‚Ä¢ Implementar regularizaci√≥n para manejar multicolinealidad
   ‚Ä¢ Validar con t√©cnicas robustas (K-fold CV)

üíº INSIGHTS PARA NEGOCIO INMOBILIARIO:
   ‚Ä¢ PRIORIZAR CALIDAD sobre cantidad de espacios
   ‚Ä¢ √ÅREA HABITABLE es m√°s importante que √°rea total del lote
   ‚Ä¢ MODERNIZACI√ìN (remodelaciones) a√±ade valor sustancial
   ‚Ä¢ GARAJES y S√ìTANOS son altamente valorados por compradores
   ‚Ä¢ EDAD DE LA CASA importa, pero puede mitigarse con renovaciones

üìä M√âTRICAS DE CALIDAD DEL AN√ÅLISIS:
   ‚Ä¢ Variables analizadas: {len(numeric_data.columns)}
   ‚Ä¢ Observaciones: {len(data_cleaned):,}
   ‚Ä¢ Cobertura de varianza (Top 10): {sum([corr_matrix['SalePrice'][f]**2 for f in top10_features]):.1%}
   ‚Ä¢ Datos limpios: {(1 - data_cleaned.isnull().sum().sum()/(len(data_cleaned)*len(data_cleaned.columns)))*100:.1f}%
""")

print("\nüéâ ¬°AN√ÅLISIS COMPLETO DE CORRELACIONES TERMINADO!")
print("‚úÖ Todos los problemas del 1 al 7 resueltos exitosamente")
print(f"üìä Dataset: {source}")
print(f"üîó URL original: https://www.kaggle.com/c/house-prices-advanced-regression-techniques")

print("\n" + "="*80)
print("üöÄ NOTEBOOK COMPLETO - READY FOR MACHINE LEARNING!")
print("="*80)
```
