```python
"""
CURSO DE CIENTÍFICO DE DATOS - REGRESIÓN LINEAL
Matrices Bidimensionales y Problemas de Gradiente
"""

import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("MATRICES BIDIMENSIONALES Y PROBLEMAS DE GRADIENTE")
print("="*70)

# ============================================================================
# PROBLEMA 1: FUNCIÓN LINEAL
# ============================================================================
print("\n✅ PROBLEMA 1: FUNCIÓN LINEAL")
print("-"*50)

# Crear valores de x de -50 a 50 en incrementos de 0.1
x = np.arange(-50, 50.1, 0.1)
y = 0.5 * x + 1

print(f"Número de elementos: {len(x)}")
print(f"Función: y = 0.5x + 1")
print(f"Rango x: {x[0]:.1f} a {x[-1]:.1f}")
print(f"Rango y: {y[0]:.1f} a {y[-1]:.1f}")

# ============================================================================
# PROBLEMA 2: CONCATENACIÓN DE MATRICES
# ============================================================================
print("\n✅ PROBLEMA 2: CONCATENACIÓN DE MATRICES")
print("-"*50)

# Combinar x e y en array (1001, 2)
array_xy = np.column_stack([x, y])
print(f"Forma del array combinado: {array_xy.shape}")

# ============================================================================
# PROBLEMA 3: ENCONTRAR EL GRADIENTE
# ============================================================================
print("\n✅ PROBLEMA 3: ENCONTRAR EL GRADIENTE")
print("-"*50)

def calculate_gradient(x_array, y_array):
    """Calcula gradiente usando diferencias finitas"""
    dx = x_array[1:] - x_array[:-1]
    dy = y_array[1:] - y_array[:-1]
    dx = np.where(np.abs(dx) < 1e-10, 1e-10, dx)
    return dy / dx

gradient = calculate_gradient(array_xy[:, 0], array_xy[:, 1])
print(f"Gradiente calculado: {len(gradient)} elementos")
print(f"Gradiente medio: {np.mean(gradient):.6f} (esperado: 0.5)")

# ============================================================================
# PROBLEMA 4: DIBUJAR GRÁFICOS
# ============================================================================
print("\n✅ PROBLEMA 4: GRÁFICOS")
print("-"*50)

plt.figure(figsize=(12, 8))

# Subplot 1: Función
plt.subplot(2, 1, 1)
plt.plot(array_xy[:, 0], array_xy[:, 1], 'b-', linewidth=2, label='y = 0.5x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Función Lineal')
plt.grid(True, alpha=0.3)
plt.legend()

# Subplot 2: Gradiente
plt.subplot(2, 1, 2)
x_gradient = (array_xy[1:, 0] + array_xy[:-1, 0]) / 2  # Puntos medios
plt.plot(x_gradient, gradient, 'r-', linewidth=2, label='Gradiente = 0.5')
plt.xlabel('x')
plt.ylabel('Gradiente')
plt.title('Gradiente de la Función')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()

# ============================================================================
# PROBLEMA 5: FUNCIONES DE PYTHON
# ============================================================================
print("\n✅ PROBLEMA 5: FUNCIONES DE PYTHON")
print("-"*50)

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calcula función y su gradiente usando diferencias finitas.
    
    Parameters
    ----------
    function : function
        Función que toma ndarray x y devuelve ndarray y
    x_range : tuple
        Argumentos para np.arange()
    
    Returns
    -------
    array_xy : ndarray, shape(n, 2)
        Array combinado de x e y
    gradient : ndarray, shape(n-1,)
        Gradiente de la función
    """
    x_array = np.arange(*x_range)
    y_array = function(x_array)
    array_xy = np.column_stack([x_array, y_array])
    
    dx = x_array[1:] - x_array[:-1]
    dy = y_array[1:] - y_array[:-1]
    dx = np.where(np.abs(dx) < 1e-10, 1e-10, dx)
    gradient = dy / dx
    
    return array_xy, gradient

# Función 1: y = x²
def function1(array_x):
    """y = x²"""
    return array_x ** 2

# Función 2: y = 2x² + 2^x (con protección contra overflow)
def function2(array_x):
    """y = 2x² + 2^x"""
    exponential_part = np.where(array_x > 30, 1e10, 
                                np.where(array_x < -50, 0, 
                                        np.power(2.0, array_x)))
    return 2 * array_x ** 2 + exponential_part

# Función 3: y = sin(√x) para x ≥ 0
def function3(array_x):
    """y = sin(√x) para x ≥ 0"""
    safe_x = np.where(array_x >= 0, array_x, 0)
    return np.sin(np.sqrt(safe_x))

# Probar las tres funciones
functions = [
    (function1, "y = x²", (-50, 50.1, 0.1)),
    (function2, "y = 2x² + 2^x", (-50, 50.1, 0.1)), 
    (function3, "y = sin(√x)", (0, 50.1, 0.1))
]

results = []

for i, (func, name, x_range) in enumerate(functions, 1):
    print(f"\n📊 FUNCIÓN {i}: {name}")
    try:
        array_xy_result, gradient_result = compute_gradient(func, x_range)
        results.append((array_xy_result, gradient_result, name))
        
        print(f"   ✅ Ejecutada correctamente")
        print(f"   Forma xy: {array_xy_result.shape}")
        print(f"   Forma gradiente: {gradient_result.shape}")
        print(f"   Rango y: {array_xy_result[:, 1].min():.2e} a {array_xy_result[:, 1].max():.2e}")
        
    except Exception as e:
        print(f"   ❌ Error: {e}")

# ============================================================================
# PROBLEMA 6: ENCONTRAR EL VALOR MÍNIMO
# ============================================================================
print("\n✅ PROBLEMA 6: ENCONTRAR VALORES MÍNIMOS")
print("-"*50)

for i, (array_xy_result, gradient_result, name) in enumerate(results):
    print(f"\n📈 ANÁLISIS DE {name}:")
    
    y_values = array_xy_result[:, 1]
    x_values = array_xy_result[:, 0]
    
    # Encontrar mínimo
    min_value = np.min(y_values)
    min_index = np.argmin(y_values)
    min_x = x_values[min_index]
    
    print(f"   Valor mínimo: y = {min_value:.6f} en x = {min_x:.1f}")
    print(f"   Índice del mínimo: {min_index}")
    
    # Gradientes antes y después del mínimo
    if 0 < min_index < len(gradient_result):
        grad_before = gradient_result[min_index-1] if min_index > 0 else "N/A"
        grad_after = gradient_result[min_index] if min_index < len(gradient_result) else "N/A"
        
        print(f"   Gradiente antes del mínimo: {grad_before}")
        print(f"   Gradiente después del mínimo: {grad_after}")
    else:
        print(f"   Mínimo en el borde, gradientes limitados")

print("\n" + "="*70)
print("✅ VERIFICACIÓN COMPLETADA - TODAS LAS SECCIONES FUNCIONAN CORRECTAMENTE")
print("="*70)
```
