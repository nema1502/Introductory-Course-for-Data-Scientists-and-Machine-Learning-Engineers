```python
"""
CURSO DE CIENTÍFICO DE DATOS - REGRESIÓN LINEAL
Problema del Descenso del Monte Fuji
Introducción al Descenso de Gradiente

Objetivo:
- Familiarizarse con la biblioteca NumPy
- Aprender cómo funciona el descenso de gradiente
- Simular el descenso de una montaña usando optimización
"""

import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("PROBLEMA DEL DESCENSO DEL MONTE FUJI")
print("="*70)

# ============================================================================
# CREACIÓN DE DATOS SINTÉTICOS DEL MONTE FUJI
# ============================================================================
def create_fuji_data():
    """Crea datos sintéticos que simulan el perfil de elevación del Monte Fuji"""
    n_points = 300
    point_numbers = np.arange(n_points)
    latitude = np.linspace(35.2, 35.4, n_points)
    longitude = np.linspace(138.6, 138.8, n_points)
    
    # Crear perfil de elevación realista del Monte Fuji
    x = np.linspace(0, 4*np.pi, n_points)
    base_elevation = 500 + 200 * np.sin(x/2)**2
    
    # Pico principal del Monte Fuji (punto 136)
    peak_center = 136
    peak_width = 30
    peak_height = 3776
    peak_factor = np.exp(-((point_numbers - peak_center) / peak_width)**2)
    peak_elevation = peak_height * peak_factor
    
    # Colinas menores
    small_peaks = 300 * (np.sin(x/3)**2 * np.exp(-((point_numbers - 50) / 40)**2) + 
                         0.5 * np.sin(x/1.5)**2 * np.exp(-((point_numbers - 220) / 35)**2))
    
    elevation = np.maximum(base_elevation + peak_elevation + small_peaks, 50)
    distance = np.cumsum(np.concatenate(([0], np.ones(n_points-1) * 100)))
    
    return np.column_stack([point_numbers, latitude, longitude, elevation, distance])

# Crear y configurar datos
fuji = create_fuji_data()
np.set_printoptions(suppress=True)

print(f"✅ Datos creados: {fuji.shape}")
print(f"Rango de elevación: {fuji[:, 3].min():.1f}m a {fuji[:, 3].max():.1f}m")
print(f"Punto más alto: {np.argmax(fuji[:, 3])} (elevación: {fuji[:, 3].max():.1f}m)")
```

```python
# ============================================================================
# PROBLEMA 1: VISUALIZACIÓN DE DATOS
# ============================================================================
print("\n✅ PROBLEMA 1: VISUALIZACIÓN DE DATOS")

plt.figure(figsize=(12, 6))
plt.plot(fuji[:, 0], fuji[:, 3], 'b-', linewidth=2)
plt.xlabel('Número de Punto')
plt.ylabel('Elevación (m)')
plt.title('Perfil de Elevación del Monte Fuji')
plt.grid(True, alpha=0.3)
plt.axhline(y=fuji[:, 3].max(), color='r', linestyle='--', alpha=0.5, label=f'Pico: {fuji[:, 3].max():.0f}m')
plt.legend()
plt.show()

print("📊 Sección transversal del Monte Fuji visualizada")
```

```python
# ============================================================================
# PROBLEMA 2: FUNCIÓN PARA CALCULAR GRADIENTE EN UN PUNTO
# ============================================================================
print("\n✅ PROBLEMA 2: FUNCIÓN PARA CALCULAR GRADIENTE")

def calculate_gradient_at_point(point_number, fuji_data):
    """
    Calcula el gradiente en un punto específico usando el punto anterior
    Fórmula: gradiente = (y_actual - y_anterior) / (x_actual - x_anterior)
    """
    if point_number <= 0:
        raise ValueError("El número de punto debe ser mayor que 0")
    if point_number >= len(fuji_data):
        raise ValueError(f"El número de punto debe ser menor que {len(fuji_data)}")
    
    current_point = int(point_number)
    previous_point = current_point - 1
    
    x_current = fuji_data[current_point, 0]
    y_current = fuji_data[current_point, 3]
    x_previous = fuji_data[previous_point, 0]
    y_previous = fuji_data[previous_point, 3]
    
    dx = x_current - x_previous
    dy = y_current - y_previous
    
    if abs(dx) < 1e-10:
        dx = 1e-10
    
    return dy / dx

# Probar función
test_points = [50, 100, 136, 150, 200, 250]
print("Punto  Elevación  Gradiente")
print("-" * 30)
for point in test_points:
    grad = calculate_gradient_at_point(point, fuji)
    elevation = fuji[point, 3]
    print(f"{point:3d}   {elevation:7.1f}   {grad:8.2f}")
```

```python
# ============================================================================
# PROBLEMA 3: FUNCIÓN PARA CALCULAR PUNTO DE DESTINO
# ============================================================================
print("\n✅ PROBLEMA 3: FUNCIÓN PARA CALCULAR PUNTO DE DESTINO")

def calculate_next_point(current_point, fuji_data, learning_rate=0.2):
    """
    Calcula el siguiente punto usando: próximo_punto = punto_actual - α × gradiente_actual
    """
    try:
        gradient = calculate_gradient_at_point(current_point, fuji_data)
    except ValueError as e:
        raise ValueError(f"No se puede calcular gradiente: {e}")
    
    next_point_float = current_point - learning_rate * gradient
    next_point = int(round(next_point_float))
    
    if next_point <= 0 or next_point >= len(fuji_data):
        raise ValueError(f"Punto de destino {next_point} fuera del rango válido")
    
    return next_point

# Demostrar efecto de diferentes tasas de aprendizaje
test_point = 150
gradient = calculate_gradient_at_point(test_point, fuji)

print(f"Punto de prueba: {test_point}, Gradiente: {gradient:.2f}")
print("Learning Rate  Next Point  Movement")
print("-" * 35)
for lr in [0.01, 0.1, 0.2, 0.5, 1.0]:
    next_pt = calculate_next_point(test_point, fuji, lr)
    movement = next_pt - test_point
    print(f"{lr:12.2f}    {next_pt:7d}   {movement:+4d}")
```

```python
# ============================================================================
# PROBLEMA 4: FUNCIÓN PARA BAJAR LA MONTAÑA (VERSIÓN CORREGIDA)
# ============================================================================
print("\n✅ PROBLEMA 4: FUNCIÓN PARA BAJAR LA MONTAÑA")

def descend_mountain(start_point, fuji_data, learning_rate=0.2, max_iterations=1000):
    """Simula el descenso de la montaña usando descenso de gradiente"""
    path = [start_point]
    elevations = [fuji_data[start_point, 3]]
    gradients = []
    
    current_point = start_point
    iteration = 0
    
    print(f"🗻 Iniciando desde punto {start_point} (elevación: {fuji_data[start_point, 3]:.1f}m)")
    
    while iteration < max_iterations:
        try:
            gradient = calculate_gradient_at_point(current_point, fuji_data)
            gradients.append(gradient)
            
            next_point = calculate_next_point(current_point, fuji_data, learning_rate)
            
            # Verificar convergencia
            if next_point == current_point:
                print(f"   ✅ Convergencia en iteración {iteration + 1}")
                break
            
            # Verificar si hemos visitado este punto recientemente (evitar ciclos)
            if len(path) > 5 and next_point in path[-5:]:
                print(f"   ⚠️  Ciclo detectado, terminando en iteración {iteration + 1}")
                break
            
            current_point = next_point
            path.append(current_point)
            elevations.append(fuji_data[current_point, 3])
            
            iteration += 1
            
        except ValueError as e:
            print(f"   ❌ Error: {e}")
            break
    
    print(f"   📊 Descenso completado: {elevations[0]:.1f}m → {elevations[-1]:.1f}m")
    return path, elevations, gradients

# Probar con punto inicial diferente al 136 para evitar convergencia inmediata
start_point = 140  # Punto en la ladera, no en el pico exacto
path, elevations, gradients = descend_mountain(start_point, fuji, learning_rate=0.2)

print(f"\nCamino: {path[:10]}..." if len(path) > 10 else f"\nCamino completo: {path}")
```

```python
# ============================================================================
# PROBLEMA 5: VISUALIZACIÓN DEL PROCESO DE DESCENSO
# ============================================================================
print("\n✅ PROBLEMA 5: VISUALIZACIÓN DEL PROCESO DE DESCENSO")

def visualize_descent(fuji_data, path, elevations, gradients, title="Proceso de Descenso"):
    """Visualiza el proceso de descenso del gradiente"""
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # 1. Perfil completo con camino marcado
    ax1.plot(fuji_data[:, 0], fuji_data[:, 3], 'b-', alpha=0.7, label='Monte Fuji')
    ax1.plot(path, [fuji_data[p, 3] for p in path], 'ro-', markersize=4, label='Camino')
    ax1.plot(path[0], fuji_data[path[0], 3], 'go', markersize=8, label='Inicio')
    ax1.plot(path[-1], fuji_data[path[-1], 3], 'ro', markersize=8, label='Final')
    ax1.set_xlabel('Punto')
    ax1.set_ylabel('Elevación (m)')
    ax1.set_title('Perfil con Camino de Descenso')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # 2. Zoom al área de descenso
    min_point, max_point = min(path), max(path)
    buffer = max(20, (max_point - min_point) // 4)
    zoom_start = max(0, min_point - buffer)
    zoom_end = min(len(fuji_data), max_point + buffer)
    
    zoom_range = range(zoom_start, zoom_end)
    ax2.plot(zoom_range, [fuji_data[i, 3] for i in zoom_range], 'b-', alpha=0.7)
    ax2.plot(path, elevations, 'ro-', markersize=6, linewidth=2, label='Camino')
    ax2.set_xlabel('Punto')
    ax2.set_ylabel('Elevación (m)')
    ax2.set_title('Zoom del Área de Descenso')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # 3. Evolución de la elevación
    iterations = range(len(elevations))
    ax3.plot(iterations, elevations, 'g-o', markersize=4)
    ax3.set_xlabel('Iteración')
    ax3.set_ylabel('Elevación (m)')
    ax3.set_title('Evolución de la Elevación')
    ax3.grid(True, alpha=0.3)
    
    # 4. Evolución del gradiente
    if gradients:
        ax4.plot(range(len(gradients)), gradients, 'r-o', markersize=4)
        ax4.axhline(y=0, color='k', linestyle='--', alpha=0.5)
        ax4.set_xlabel('Iteración')
        ax4.set_ylabel('Gradiente')
        ax4.set_title('Evolución del Gradiente')
        ax4.grid(True, alpha=0.3)
    
    plt.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# Visualizar el descenso anterior
visualize_descent(fuji, path, elevations, gradients, f"Descenso desde Punto {start_point}")
```

```python
# ============================================================================
# PROBLEMA 6: CAMBIAR VALOR INICIAL
# ============================================================================
print("\n✅ PROBLEMA 6: CAMBIAR VALOR INICIAL")

def test_multiple_starting_points(fuji_data, starting_points, learning_rate=0.2):
    """Prueba el descenso desde múltiples puntos iniciales"""
    results = {}
    
    for start in starting_points:
        print(f"\n🧪 Probando desde punto {start}")
        try:
            path, elevations, gradients = descend_mountain(start, fuji_data, learning_rate, max_iterations=100)
            results[start] = {
                'path': path,
                'elevations': elevations,
                'gradients': gradients,
                'final_point': path[-1],
                'final_elevation': elevations[-1],
                'descent': elevations[0] - elevations[-1]
            }
        except Exception as e:
            print(f"   ❌ Error: {e}")
            results[start] = None
    
    return results

# Probar diferentes puntos iniciales
starting_points = [50, 80, 120, 140, 160, 200, 250]
results = test_multiple_starting_points(fuji, starting_points)

print(f"\n📊 RESUMEN DE RESULTADOS:")
print("Inicio  Final  Descenso  Iteraciones")
print("-" * 38)
for start in starting_points:
    if results[start]:
        r = results[start]
        iterations = len(r['path']) - 1
        print(f"{start:6d}  {r['final_point']:5d}  {r['descent']:8.1f}  {iterations:11d}")
    else:
        print(f"{start:6d}  Error")
```

```python
# ============================================================================
# PROBLEMA 7: VISUALIZACIÓN PARA MÚLTIPLES VALORES INICIALES
# ============================================================================
print("\n✅ PROBLEMA 7: VISUALIZACIÓN MÚLTIPLES INICIOS")

def visualize_multiple_descents(fuji_data, results_dict, title="Descensos desde Múltiples Puntos"):
    """Visualiza múltiples descensos en el mismo gráfico"""
    
    plt.figure(figsize=(15, 8))
    
    # Colores para diferentes caminos
    colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink']
    
    # Dibujar perfil del Monte Fuji
    plt.plot(fuji_data[:, 0], fuji_data[:, 3], 'k-', alpha=0.3, linewidth=2, label='Monte Fuji')
    
    # Dibujar cada camino de descenso
    color_idx = 0
    for start_point, result in results_dict.items():
        if result is not None:
            path = result['path']
            elevations = result['elevations']
            color = colors[color_idx % len(colors)]
            
            # Camino completo
            plt.plot(path, elevations, 'o-', color=color, markersize=3, 
                    alpha=0.7, label=f'Inicio: {start_point}')
            
            # Marcar puntos inicial y final
            plt.plot(path[0], elevations[0], 'o', color=color, markersize=8, 
                    markeredgecolor='black', markeredgewidth=1)
            plt.plot(path[-1], elevations[-1], 's', color=color, markersize=8,
                    markeredgecolor='black', markeredgewidth=1)
            
            color_idx += 1
    
    plt.xlabel('Número de Punto')
    plt.ylabel('Elevación (m)')
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    # Gráfico de convergencia
    plt.figure(figsize=(12, 6))
    
    for start_point, result in results_dict.items():
        if result is not None:
            elevations = result['elevations']
            iterations = range(len(elevations))
            plt.plot(iterations, elevations, 'o-', label=f'Inicio: {start_point}', 
                    markersize=4, alpha=0.7)
    
    plt.xlabel('Iteración')
    plt.ylabel('Elevación (m)')
    plt.title('Convergencia de Elevación por Iteración')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Visualizar todos los descensos
visualize_multiple_descents(fuji, results, "Descenso de Gradiente desde Múltiples Puntos Iniciales")
```

```python
# ============================================================================
# PROBLEMA 8: CAMBIAR HIPERPARÁMETROS (PROBLEMA AVANZADO)
# ============================================================================
print("\n✅ PROBLEMA 8: EFECTO DE DIFERENTES HIPERPARÁMETROS")

def compare_learning_rates(fuji_data, start_point, learning_rates):
    """Compara el efecto de diferentes tasas de aprendizaje"""
    results = {}
    
    for lr in learning_rates:
        print(f"\n🧪 Probando learning rate = {lr}")
        try:
            path, elevations, gradients = descend_mountain(start_point, fuji_data, lr, max_iterations=200)
            results[lr] = {
                'path': path,
                'elevations': elevations,
                'gradients': gradients,
                'iterations': len(path) - 1,
                'final_elevation': elevations[-1],
                'convergence_speed': len(path) - 1
            }
        except Exception as e:
            print(f"   ❌ Error con lr={lr}: {e}")
            results[lr] = None
    
    return results

# Probar diferentes learning rates
start_point = 140
learning_rates = [0.01, 0.1, 0.2, 0.5, 0.8]
lr_results = compare_learning_rates(fuji, start_point, learning_rates)

# Visualizar comparación de learning rates
def visualize_learning_rates(fuji_data, start_point, lr_results):
    """Visualiza el efecto de diferentes learning rates"""
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    colors = ['blue', 'green', 'red', 'orange', 'purple']
    
    # 1. Caminos en el perfil
    ax1.plot(fuji_data[:, 0], fuji_data[:, 3], 'k-', alpha=0.3, linewidth=2, label='Monte Fuji')
    
    for i, (lr, result) in enumerate(lr_results.items()):
        if result is not None:
            path = result['path']
            elevations = result['elevations']
            color = colors[i % len(colors)]
            ax1.plot(path, elevations, 'o-', color=color, markersize=3, 
                    alpha=0.7, label=f'α = {lr}')
    
    ax1.set_xlabel('Punto')
    ax1.set_ylabel('Elevación (m)')
    ax1.set_title('Caminos con Diferentes Learning Rates')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # 2. Convergencia
    for i, (lr, result) in enumerate(lr_results.items()):
        if result is not None:
            elevations = result['elevations']
            iterations = range(len(elevations))
            color = colors[i % len(colors)]
            ax2.plot(iterations, elevations, 'o-', color=color, 
                    markersize=4, alpha=0.7, label=f'α = {lr}')
    
    ax2.set_xlabel('Iteración')
    ax2.set_ylabel('Elevación (m)')
    ax2.set_title('Convergencia por Learning Rate')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # 3. Análisis de iteraciones vs learning rate
    lrs = []
    iterations = []
    for lr, result in lr_results.items():
        if result is not None:
            lrs.append(lr)
            iterations.append(result['iterations'])
    
    ax3.bar(range(len(lrs)), iterations, color=colors[:len(lrs)])
    ax3.set_xlabel('Learning Rate')
    ax3.set_ylabel('Número de Iteraciones')
    ax3.set_title('Iteraciones para Convergencia')
    ax3.set_xticks(range(len(lrs)))
    ax3.set_xticklabels([f'{lr:.2f}' for lr in lrs])
    ax3.grid(True, alpha=0.3)
    
    # 4. Elevación final vs learning rate
    final_elevs = []
    for lr, result in lr_results.items():
        if result is not None:
            final_elevs.append(result['final_elevation'])
    
    ax4.bar(range(len(lrs)), final_elevs, color=colors[:len(lrs)])
    ax4.set_xlabel('Learning Rate')
    ax4.set_ylabel('Elevación Final (m)')
    ax4.set_title('Elevación Final por Learning Rate')
    ax4.set_xticks(range(len(lrs)))
    ax4.set_xticklabels([f'{lr:.2f}' for lr in lrs])
    ax4.grid(True, alpha=0.3)
    
    plt.suptitle(f'Análisis de Learning Rates (Inicio: Punto {start_point})', fontsize=16)
    plt.tight_layout()
    plt.show()

visualize_learning_rates(fuji, start_point, lr_results)

print(f"\n📊 RESUMEN FINAL:")
print("Learning Rate  Iteraciones  Elevación Final")
print("-" * 44)
for lr, result in lr_results.items():
    if result is not None:
        print(f"{lr:12.2f}  {result['iterations']:11d}  {result['final_elevation']:13.1f}")

print(f"\n🎉 ¡DESCENSO DEL MONTE FUJI COMPLETADO!")
print("✅ Todos los problemas implementados y verificados")
```
