```python
"""
CURSO DE CIENT√çFICO DE DATOS - REGRESI√ìN LINEAL
An√°lisis de Datos de Iris
Introducci√≥n al An√°lisis de Datos con Pandas

Objetivo:
- Comprender el flujo b√°sico del an√°lisis de datos
- Familiarizarse con la biblioteca de an√°lisis de datos Pandas
- Explorar el famoso conjunto de datos Iris de Fisher
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings('ignore')

# Configurar estilo de visualizaci√≥n
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (10, 6)

print("="*70)
print("AN√ÅLISIS DE DATOS DE IRIS DE FISHER")
print("="*70)
```

```python
# ============================================================================
# 3. PREPARACI√ìN DEL CONJUNTO DE DATOS
# ============================================================================
print("‚úÖ SECCI√ìN 3: PREPARACI√ìN DEL CONJUNTO DE DATOS")

# Cargar el conjunto de datos Iris
iris = load_iris()

print(f"üìä Dataset Iris cargado:")
print(f"   ‚Ä¢ Muestras: {iris.data.shape[0]}")
print(f"   ‚Ä¢ Caracter√≠sticas: {iris.data.shape[1]}")
print(f"   ‚Ä¢ Clases: {iris.target_names}")

# Crear DataFrame X (variables explicativas)
X = pd.DataFrame(iris.data, columns=iris.feature_names)
column_mapping = {
    'sepal length (cm)': 'longitud_sepalo',
    'sepal width (cm)': 'ancho_sepalo', 
    'petal length (cm)': 'longitud_petalo',
    'petal width (cm)': 'ancho_petalo'
}
X = X.rename(columns=column_mapping)

# Crear DataFrame y (variable objetivo)
y = pd.DataFrame(iris.target, columns=['especies'])

# Combinar DataFrames
df = pd.concat([X, y], axis=1)
species_mapping = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['nombres_especies'] = df['especies'].map(species_mapping)

print(f"‚úÖ DataFrame combinado creado: {df.shape}")
print(df.head())
```

```python
# ============================================================================
# 4. DESCRIPCI√ìN GENERAL DE LOS DATOS
# ============================================================================
print("\n‚úÖ SECCI√ìN 4: DESCRIPCI√ìN GENERAL DE LOS DATOS")

print("üìã 1. Primeras 4 muestras:")
print(df.head(4))

print(f"\nüìä 2. Tipos de datos:")
print(df.dtypes)

print(f"\nüîç 3. Informaci√≥n del DataFrame:")
df.info()

print(f"\nüìà 4. Conteo por especies:")
print(df['nombres_especies'].value_counts())

print(f"\n‚ùì 5. Valores faltantes:")
missing = df.isnull().sum()
print(missing)
print(f"Total: {missing.sum()} ({'‚úÖ Perfecto' if missing.sum() == 0 else '‚ö†Ô∏è Hay faltantes'})")

print(f"\nüìä 6. Estad√≠sticas descriptivas:")
print(df.describe())
```

```python
# ============================================================================
# 5. EXPLORACI√ìN DE CARACTER√çSTICAS - EXTRACCI√ìN DE DATOS
# ============================================================================
print("\n‚úÖ SECCI√ìN 5: EXPLORACI√ìN DE CARACTER√çSTICAS")

# 1. Extraer columna 'ancho_sepalo' de dos maneras
print("1Ô∏è‚É£ Extracci√≥n de 'ancho_sepalo':")
sepal_width1 = df.ancho_sepalo  # M√©todo 1
sepal_width2 = df['ancho_sepalo']  # M√©todo 2
print(f"M√©todo 1 (punto): {sepal_width1.head().values}")
print(f"M√©todo 2 (corchetes): {sepal_width2.head().values}")
print(f"¬øSon iguales? {sepal_width1.equals(sepal_width2)}")

# 2. Extraer filas 50-99
print(f"\n2Ô∏è‚É£ Filas 50-99:")
rows_50_99 = df.iloc[50:100]
print(f"Forma: {rows_50_99.shape}")
print(f"Especies: {rows_50_99['nombres_especies'].unique()}")

# 3. Longitud p√©talo de filas 50-99  
print(f"\n3Ô∏è‚É£ Longitud p√©talo (filas 50-99):")
petal_length_50_99 = df.loc[50:99, 'longitud_petalo']
print(f"Estad√≠sticas: min={petal_length_50_99.min()}, max={petal_length_50_99.max()}")

# 4. Datos donde ancho_petalo = 0.2
print(f"\n4Ô∏è‚É£ Muestras con ancho_petalo = 0.2:")
petal_width_02 = df[df['ancho_petalo'] == 0.2]
print(f"Cantidad: {len(petal_width_02)}")
print(f"Especies: {petal_width_02['nombres_especies'].unique()}")
```

```python
# ============================================================================
# VISUALIZACIONES - GR√ÅFICO CIRCULAR
# ============================================================================
print("\nüìä VISUALIZACIONES")

# Gr√°fico circular de especies
plt.figure(figsize=(8, 8))
species_counts = df['nombres_especies'].value_counts()
colors = ['#FF9999', '#66B2FF', '#99FF99']

plt.pie(species_counts.values, 
        labels=species_counts.index,
        autopct='%1.1f%%',
        colors=colors,
        startangle=90)
plt.title('Distribuci√≥n de Especies de Iris', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()

print(f"‚úÖ Gr√°fico circular: Distribuci√≥n perfectamente equilibrada (33.3% cada especie)")
```

```python
# ============================================================================
# DIAGRAMAS DE CAJA POR CARACTER√çSTICA
# ============================================================================
print("\nüìä DIAGRAMAS DE CAJA")

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
features = ['longitud_sepalo', 'ancho_sepalo', 'longitud_petalo', 'ancho_petalo']
titles = ['Longitud del S√©palo', 'Ancho del S√©palo', 'Longitud del P√©talo', 'Ancho del P√©talo']

for i, (feature, title) in enumerate(zip(features, titles)):
    row, col = i // 2, i % 2
    
    # Preparar datos para boxplot
    data_by_species = [df[df['nombres_especies'] == species][feature].values 
                       for species in ['setosa', 'versicolor', 'virginica']]
    
    axes[row, col].boxplot(data_by_species, labels=['Setosa', 'Versicolor', 'Virginica'])
    axes[row, col].set_title(title, fontweight='bold')
    axes[row, col].set_ylabel('Cent√≠metros')
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("‚úÖ Diagramas de caja: Muestran clara separaci√≥n entre especies")
```

```python
# ============================================================================
# DIAGRAMAS DE VIOL√çN
# ============================================================================
print("\nüéª DIAGRAMAS DE VIOL√çN")

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for i, (feature, title) in enumerate(zip(features, titles)):
    row, col = i // 2, i % 2
    
    sns.violinplot(data=df, x='nombres_especies', y=feature, ax=axes[row, col])
    axes[row, col].set_title(title, fontweight='bold')
    axes[row, col].set_xlabel('Especie')
    axes[row, col].set_ylabel('Cent√≠metros')

plt.tight_layout()
plt.show()

print("‚úÖ Diagramas de viol√≠n: Revelan la distribuci√≥n detallada de cada caracter√≠stica")
```

```python
# ============================================================================
# DIAGRAMAS DE DISPERSI√ìN - TODAS LAS COMBINACIONES
# ============================================================================
print("\nüìä DIAGRAMAS DE DISPERSI√ìN - COMBINACIONES DE CARACTER√çSTICAS")

# Crear todas las combinaciones (6 en total)
combinations = [
    ('longitud_sepalo', 'ancho_sepalo'),
    ('longitud_sepalo', 'longitud_petalo'),
    ('longitud_sepalo', 'ancho_petalo'),
    ('ancho_sepalo', 'longitud_petalo'),
    ('ancho_sepalo', 'ancho_petalo'),
    ('longitud_petalo', 'ancho_petalo')
]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
colors = ['red', 'blue', 'green']
species_list = ['setosa', 'versicolor', 'virginica']

for i, (feature1, feature2) in enumerate(combinations):
    row, col = i // 3, i % 3
    
    for j, species in enumerate(species_list):
        species_data = df[df['nombres_especies'] == species]
        axes[row, col].scatter(species_data[feature1], species_data[feature2], 
                              c=colors[j], label=species, alpha=0.7, s=50)
    
    axes[row, col].set_xlabel(feature1.replace('_', ' ').title())
    axes[row, col].set_ylabel(feature2.replace('_', ' ').title())
    axes[row, col].set_title(f'{feature1} vs {feature2}', fontweight='bold')
    axes[row, col].legend()
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("‚úÖ Diagramas de dispersi√≥n: Muestran separaci√≥n clara entre especies")
```

```python
# ============================================================================
# MATRIZ DE DIAGRAMAS DE DISPERSI√ìN (PAIRPLOT)
# ============================================================================
print("\nüìä MATRIZ DE DIAGRAMAS DE DISPERSI√ìN")

# Crear pairplot con seaborn
plt.figure(figsize=(12, 10))
pairplot = sns.pairplot(df[['longitud_sepalo', 'ancho_sepalo', 'longitud_petalo', 
                           'ancho_petalo', 'nombres_especies']], 
                       hue='nombres_especies', 
                       diag_kind='hist',
                       markers=['o', 's', '^'])

pairplot.fig.suptitle('Matriz de Diagramas de Dispersi√≥n - Dataset Iris', 
                     y=1.02, fontsize=16, fontweight='bold')
plt.show()

print("‚úÖ Pairplot: Vista comprehensiva de todas las relaciones entre caracter√≠sticas")
```

```python
# ============================================================================
# MATRIZ DE CORRELACI√ìN
# ============================================================================
print("\nüìä AN√ÅLISIS DE CORRELACI√ìN")

# Calcular matriz de correlaci√≥n (solo caracter√≠sticas num√©ricas)
numeric_features = ['longitud_sepalo', 'ancho_sepalo', 'longitud_petalo', 'ancho_petalo']
correlation_matrix = df[numeric_features].corr()

print("üìä Matriz de Correlaci√≥n:")
print(correlation_matrix.round(3))

# Crear heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='RdYlBu_r', 
            vmin=-1, vmax=1,
            center=0,
            square=True,
            fmt='.3f')
plt.title('Mapa de Calor - Matriz de Correlaci√≥n', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nüîç AN√ÅLISIS DE CORRELACIONES:")
print("‚Ä¢ Longitud p√©talo ‚Üî Ancho p√©talo: Correlaci√≥n ALTA (0.963)")
print("‚Ä¢ Longitud s√©palo ‚Üî Longitud p√©talo: Correlaci√≥n ALTA (0.872)")  
print("‚Ä¢ Longitud s√©palo ‚Üî Ancho p√©talo: Correlaci√≥n MODERADA (0.818)")
print("‚Ä¢ Ancho s√©palo ‚Üî Longitud p√©talo: Correlaci√≥n NEGATIVA (-0.428)")
```

```python
# ============================================================================
# EXPLICACI√ìN DE RESULTADOS
# ============================================================================
print("\n‚úÖ EXPLICACI√ìN DE LOS RESULTADOS")
print("="*70)

print("üå∏ CARACTER√çSTICAS DE CADA ESPECIE:")
print("""
SETOSA:
‚Ä¢ P√©talos m√°s peque√±os (longitud ~1.5cm, ancho ~0.2cm)
‚Ä¢ S√©palos m√°s anchos en promedio (~3.4cm)
‚Ä¢ Muy distinguible del resto
‚Ä¢ Sin solapamiento con otras especies

VERSICOLOR: 
‚Ä¢ Caracter√≠sticas intermedias entre setosa y virginica
‚Ä¢ Longitud p√©talo media (~4.3cm)
‚Ä¢ Ancho p√©talo moderado (~1.3cm)
‚Ä¢ Alg√∫n solapamiento con virginica

VIRGINICA:
‚Ä¢ P√©talos m√°s grandes (longitud ~5.6cm, ancho ~2.0cm)  
‚Ä¢ S√©palos m√°s largos (~6.6cm)
‚Ä¢ Ligero solapamiento con versicolor en algunas caracter√≠sticas
""")

print("üìä HALLAZGOS CLAVE DEL AN√ÅLISIS:")
print("""
1. SEPARABILIDAD: 
   ‚Ä¢ Setosa es completamente separable
   ‚Ä¢ Versicolor y virginica tienen alg√∫n solapamiento
   
2. CARACTER√çSTICAS M√ÅS DISCRIMINANTES:
   ‚Ä¢ Longitud y ancho del p√©talo son las m√°s √∫tiles
   ‚Ä¢ Ancho del s√©palo es menos discriminante
   
3. CORRELACIONES IMPORTANTES:
   ‚Ä¢ Caracter√≠sticas del p√©talo muy correlacionadas (0.963)
   ‚Ä¢ Longitud s√©palo-p√©talo tambi√©n correlacionadas (0.872)
   
4. DISTRIBUCIONES:
   ‚Ä¢ Dataset perfectamente balanceado (50 muestras/especie)
   ‚Ä¢ Sin valores faltantes
   ‚Ä¢ Caracter√≠sticas siguen distribuciones aproximadamente normales
""")

print("\nüéØ CONCLUSIONES PARA MACHINE LEARNING:")
print("""
‚Ä¢ Clasificaci√≥n ser√° muy exitosa debido a clara separaci√≥n
‚Ä¢ Setosa ser√° 100% clasificable
‚Ä¢ Versicolor y virginica requerir√°n m√°s cuidado
‚Ä¢ Caracter√≠sticas del p√©talo son las m√°s informativas
‚Ä¢ Reducci√≥n de dimensionalidad podr√≠a ser efectiva
""")

print(f"\nüéâ ¬°AN√ÅLISIS DE IRIS COMPLETADO!")
print("‚úÖ Todas las secciones implementadas y verificadas")
```
