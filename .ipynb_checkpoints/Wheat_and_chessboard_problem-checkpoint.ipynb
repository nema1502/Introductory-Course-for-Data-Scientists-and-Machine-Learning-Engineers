```python
"""
CURSO DE CIENTÍFICO DE DATOS - REGRESIÓN LINEAL
Problema del Trigo y el Tablero de Ajedrez
Introducción a NumPy y Visualizaciones

Objetivo:
- Familiarizarse con la biblioteca NumPy
- Aprender visualizaciones gráficas
- Entender el crecimiento exponencial
"""

import numpy as np
import matplotlib.pyplot as plt
import time

print("="*60)
print("PROBLEMA DEL TRIGO Y EL TABLERO DE AJEDREZ")
print("="*60)

# 1. TABLERO PEQUEÑO (4 CASILLAS)
print("\n1. TABLERO PEQUEÑO (4 casillas)")
print("-"*40)

# Método usando listas de Python
n_squares = 4
small_board_list = [1]
for _ in range(n_squares - 1):
    small_board_list.append(2*small_board_list[-1])
print(f"4 casillas con listas: {small_board_list}")

# Convertir a ndarray
small_board_ndarray = np.array(small_board_list)
print(f"4 casillas con ndarray: {small_board_ndarray}")

# 2. TABLERO DE AJEDREZ 2x2
print("\n2. TABLERO DE AJEDREZ 2x2")
print("-"*40)

# Crear un tablero 2x2 con la secuencia 1, 2, 4, 8
board_2x2_flat = np.array([1, 2, 4, 8])
board_2x2 = board_2x2_flat.reshape(2, 2)
print("Tablero 2x2:")
print(board_2x2)
print(f"Forma del tablero: {board_2x2.shape}")

# 3. FUNCIÓN PARA TABLEROS n×m
print("\n3. FUNCIÓN PARA TABLEROS n×m")
print("-"*40)

def create_wheat_board(n_rows, n_cols):
    """
    Crea un tablero de ajedrez n×m con granos de trigo siguiendo la secuencia 2^i
    
    Args:
        n_rows (int): Número de filas
        n_cols (int): Número de columnas
    
    Returns:
        numpy.ndarray: Tablero con la cantidad de granos en cada casilla
    """
    total_squares = n_rows * n_cols
    # Crear índices de 0 a total_squares-1
    indices = np.arange(total_squares, dtype=np.uint64)
    # Calcular 2^i para cada índice
    wheat_counts = 2**indices
    # Reshape al tamaño del tablero
    return wheat_counts.reshape(n_rows, n_cols)

# Probar con tablero 4x4
board_4x4 = create_wheat_board(4, 4)
print("Tablero 4x4:")
print(board_4x4)

# 4. TABLERO DE AJEDREZ 8x8
print("\n4. TABLERO DE AJEDREZ 8x8")
print("-"*40)

# Crear tablero 8x8
board_8x8 = create_wheat_board(8, 8)
print("Tablero 8x8 (primeras 4 filas y columnas):")
print(board_8x8[:4, :4])
print(f"Forma del tablero: {board_8x8.shape}")

# 5. NÚMERO TOTAL DE TRIGO Y ANÁLISIS
print("\n5. ANÁLISIS DEL TRIGO EN TABLERO 8x8")
print("-"*40)

total_wheat = np.sum(board_8x8)
print(f"Número total de granos de trigo: {total_wheat:,}")

# Calcular promedio por columna
column_averages = np.mean(board_8x8, axis=0)
print(f"\nPromedio de granos por columna:")
for i, avg in enumerate(column_averages):
    print(f"Columna {i+1}: {avg:,.0f}")

# 6. VISUALIZACIÓN: GRÁFICO DE BARRAS
print("\n6. VISUALIZACIÓN: GRÁFICO DE BARRAS")
print("-"*40)

plt.figure(figsize=(10, 6))
columns = np.arange(1, 9)
plt.xlabel("Columna")
plt.ylabel("Número promedio de granos")
plt.title("Promedio de Granos por Columna en Tablero 8x8")
plt.bar(columns, column_averages)
plt.yscale('log')  # Escala logarítmica para mejor visualización
plt.grid(True, alpha=0.3)
plt.show()

# 7. MAPA DE CALOR
print("\n7. MAPA DE CALOR DEL TABLERO")
print("-"*40)

plt.figure(figsize=(10, 8))
plt.xlabel("Columna")
plt.ylabel("Fila")
plt.title("Mapa de Calor: Distribución de Trigo (Escala Logarítmica)")
# Usar logaritmo para mejor visualización
plt.pcolor(np.log10(board_8x8), cmap='viridis')
plt.colorbar(label='log10(Número de granos)')
plt.show()

# 8. LA SEGUNDA MITAD DEL TABLERO DE AJEDREZ
print("\n8. COMPARACIÓN: PRIMERA vs SEGUNDA MITAD")
print("-"*40)

# Primera mitad (filas 0-3) y segunda mitad (filas 4-7)
first_half = board_8x8[:4, :]
second_half = board_8x8[4:, :]

total_first_half = np.sum(first_half)
total_second_half = np.sum(second_half)

print(f"Total en primera mitad (filas 0-3): {total_first_half:,}")
print(f"Total en segunda mitad (filas 4-7): {total_second_half:,}")

ratio = total_second_half / total_first_half
print(f"La segunda mitad tiene {ratio:,.0f} veces más granos que la primera")

# 9. OTROS MÉTODOS DE CÁLCULO
print("\n9. OTROS MÉTODOS DE CÁLCULO")
print("-"*40)

# Método con np.append()
def create_wheat_board_append(n_rows, n_cols):
    total_squares = n_rows * n_cols
    board_ndarray = np.array([1], dtype=np.uint64)
    for _ in range(total_squares - 1):
        board_ndarray = np.append(board_ndarray, 2*board_ndarray[-1])
    return board_ndarray.reshape(n_rows, n_cols)

# Método con broadcasting (más eficiente)
def create_wheat_board_broadcast(n_rows, n_cols):
    total_squares = n_rows * n_cols
    indices_of_squares = np.arange(total_squares, dtype=np.uint64)
    board_ndarray = 2**indices_of_squares
    return board_ndarray.reshape(n_rows, n_cols)

# 10. COMPARACIÓN DE RENDIMIENTO
print("\n10. COMPARACIÓN DE RENDIMIENTO")
print("-"*40)

# Medir tiempo para tablero 8x8
print("Tiempo de ejecución para tablero 8x8:")

# Método con listas
start_time = time.time()
board_list = create_wheat_board(8, 8)
time_list = time.time() - start_time
print(f"Método función original: {time_list:.6f} segundos")

# Método con append
start_time = time.time()
board_append = create_wheat_board_append(8, 8)
time_append = time.time() - start_time
print(f"Método con np.append(): {time_append:.6f} segundos")

# Método con broadcasting
start_time = time.time()
board_broadcast = create_wheat_board_broadcast(8, 8)
time_broadcast = time.time() - start_time
print(f"Método con broadcasting: {time_broadcast:.6f} segundos")

# Verificar que todos los métodos dan el mismo resultado
print(f"\n¿Todos los métodos dan el mismo resultado? {np.array_equal(board_list, board_append) and np.array_equal(board_append, board_broadcast)}")

# 11. ANÁLISIS DE TIPOS DE DATOS
print("\n11. ANÁLISIS DE TIPOS DE DATOS")
print("-"*40)

# Demostrar problema con int64
try:
    board_int64 = 2**np.arange(64, dtype=np.int64)
    print(f"Último valor con int64: {board_int64[-1]}")
    print("⚠️  Valor negativo debido a overflow de int64")
except:
    print("Error con int64")

# Solución con uint64
board_uint64 = 2**np.arange(64, dtype=np.uint64)
print(f"Último valor con uint64: {board_uint64[-1]}")
print("✅ Valor correcto con uint64")

# 12. RESUMEN DE RESULTADOS
print("\n12. RESUMEN DE RESULTADOS")
print("-"*40)
print(f"• Total de granos en tablero 8x8: {total_wheat:,}")
print(f"• Granos en la última casilla: {board_8x8[-1, -1]:,}")
print(f"• Ratio segunda mitad / primera mitad: {ratio:,.0f}")
print(f"• El método más eficiente es: broadcasting")
print("• Es importante usar uint64 para números grandes")

print("\n" + "="*60)
print("ANÁLISIS COMPLETADO")
print("="*60)
```
