# Problema de Soroshi Shinzaemon
Hay una anécdota de Soroshi Shinzaemon que es similar a la historia que llega a la luna cuando doblas el papel tratado en el problema "Usemos Python". Vamos a proceder con la programación utilizando esto como material.

### Problema 1: Número de granos de arroz en el día 100

```python
import matplotlib.pyplot as plt

n_days = 100
grains_per_day = [2**i for i in range(n_days)]  # Día 1: 2^0=1, día 100: 2^99
total_grains = []
total = 0

for grains in grains_per_day:
    total += grains
    total_grains.append(total)

print(f"Número total de granos en el día 100: {total_grains[-1]}")

# Crear gráfico
dias = list(range(1, n_days + 1))
plt.figure(figsize=(10, 6))
plt.plot(dias, grains_per_day, label='Granos por día')
plt.plot(dias, total_grains, label='Total acumulado')
plt.xlabel('Días')
plt.ylabel('Cantidad de arroz')
plt.title('Problema 1: Transición de Granos de Arroz en 100 Días')
plt.yscale('log')  # Escala logarítmica para mejor visualización
plt.legend()
plt.grid(True)
plt.show()
```

### Problema 2: Función generalizada para N días

```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """Calcula la cantidad de arroz recibido y acumulado para cada día según la anécdota de Sorori Shinzaemon.
    
    Parámetros:
    n_days (int): Número de días a calcular. Por defecto 100.
    
    Retorna:
    tuple: (list_n_grains, list_total_grains)
        - list_n_grains: Lista de granos recibidos cada día.
        - list_total_grains: Lista acumulada de granos hasta cada día.
    """
    list_n_grains = [2**i for i in range(n_days)]
    list_total_grains = []
    total = 0
    for grains in list_n_grains:
        total += grains
        list_total_grains.append(total)
    return list_n_grains, list_total_grains

# Ejemplo para 30 días
n_days = 30
grains_diarios, total_acumulado = compute_sorori_shinzaemon(n_days)
print(f"Total de granos después de {n_days} días: {total_acumulado[-1]}")

# Gráfico
dias = list(range(1, n_days + 1))
plt.figure(figsize=(10, 6))
plt.plot(dias, grains_diarios, label='Granos por día')
plt.plot(dias, total_acumulado, label='Total acumulado')
plt.xlabel('Días')
plt.ylabel('Cantidad de arroz')
plt.title(f'Transición de Granos de Arroz en {n_days} Días')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

### Problema 3: Cálculo de supervivencia

```python
def calcular_dias_sobrevivencia(total_grains, num_personas):
    """Calcula cuántos días un grupo de personas puede sobrevivir con una cantidad de arroz.
    
    Supuestos:
    - Cada persona consume 10,000 granos al día (basado en que 1 kg ≈ 50,000 granos y 200g/día/persona).
    
    Parámetros:
    total_grains (int): Total de granos disponibles.
    num_personas (int): Número de personas.
    
    Retorna:
    int: Días de supervivencia.
    """
    consumo_diario = num_personas * 10000
    if consumo_diario == 0:
        return 0  # Evita división por cero
    return total_grains // consumo_diario

# Ejemplo de uso
total_ejemplo = 10**6  # 1,000,000 granos
personas = 5
dias = calcular_dias_sobrevivencia(total_ejemplo, personas)
print(f"{personas} personas sobreviven {dias} días con {total_ejemplo} granos.")
```
