```python
"""
CURSO DE CIENTÍFICO DE DATOS - REGRESIÓN LINEAL
Creación de Datos Ficticios
Introducción a NumPy y Visualizaciones

Objetivo:
- Familiarizarse con la biblioteca NumPy
- Aprender a generar datos aleatorios con distribución normal
- Crear visualizaciones con matplotlib
- Preparar datos para aprendizaje automático
"""

import numpy as np
import matplotlib.pyplot as plt

print("="*60)
print("CREACIÓN DE DATOS FICTICIOS")
print("="*60)

# Fijar semilla para reproducibilidad
np.random.seed(0)
print("Semilla fijada en 0 para reproducibilidad\n")
```

```python
# ============================================================================
# PROBLEMA 1: Creación de números aleatorios
# ============================================================================
print("PROBLEMA 1: Creación de números aleatorios")
print("-" * 50)

# Parámetros para la distribución normal bidimensional
mean_1 = [-3, 0]  # Media (-3, 0)
cov_matrix = [[1.0, 0.8], [0.8, 1.0]]  # Matriz de covarianza

# Generar 500 números aleatorios usando np.random.multivariate_normal()
data_problem1 = np.random.multivariate_normal(mean_1, cov_matrix, 500)

print(f"Forma de los datos del Problema 1: {data_problem1.shape}")
print(f"Media real de los datos generados: {np.mean(data_problem1, axis=0)}")
print(f"Desviación estándar: {np.std(data_problem1, axis=0)}")
print(f"Primeras 5 filas:")
print(data_problem1[:5])

# Separar las dimensiones para facilitar su uso
x1, y1 = data_problem1[:, 0], data_problem1[:, 1]
```

```python
# ============================================================================
# PROBLEMA 2: Visualización mediante un diagrama de dispersión
# ============================================================================
print("\nPROBLEMA 2: Visualización mediante diagrama de dispersión")
print("-" * 50)

plt.figure(figsize=(8, 6))
plt.scatter(x1, y1, alpha=0.6, c='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Problema 2: Diagrama de Dispersión - Primer Conjunto de Datos')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.show()

print(f"Rango X: [{np.min(x1):.2f}, {np.max(x1):.2f}]")
print(f"Rango Y: [{np.min(y1):.2f}, {np.max(y1):.2f}]")
```

```python
# ============================================================================
# PROBLEMA 3: Visualización mediante histogramas
# ============================================================================
print("\nPROBLEMA 3: Visualización mediante histogramas")
print("-" * 50)

# Calcular rango común para alinear los histogramas
x_min, x_max = min(np.min(x1), np.min(y1)), max(np.max(x1), np.max(y1))
x_range = [x_min - 0.5, x_max + 0.5]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Histograma para dimensión X
ax1.hist(x1, bins=30, alpha=0.7, color='blue', edgecolor='black')
ax1.set_xlim(x_range)
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')
ax1.set_title('Histograma X (Primera Dimensión)')
ax1.grid(True, alpha=0.3)

# Histograma para dimensión Y
ax2.hist(y1, bins=30, alpha=0.7, color='green', edgecolor='black')
ax2.set_xlim(x_range)
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')
ax2.set_title('Histograma Y (Segunda Dimensión)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Ambos histogramas usan el rango: {x_range}")
```

```python
# ============================================================================
# PROBLEMA 4: Agregar datos (segundo conjunto)
# ============================================================================
print("\nPROBLEMA 4: Agregar datos")
print("-" * 50)

# Parámetros para el segundo conjunto de datos
mean_2 = [0, -3]  # Media (0, -3)
# Misma matriz de covarianza
cov_matrix = [[1.0, 0.8], [0.8, 1.0]]

# Generar 500 números aleatorios más
data_problem4 = np.random.multivariate_normal(mean_2, cov_matrix, 500)

print(f"Forma de los datos del Problema 4: {data_problem4.shape}")
print(f"Media real de los datos generados: {np.mean(data_problem4, axis=0)}")
print(f"Desviación estándar: {np.std(data_problem4, axis=0)}")
print(f"Primeras 5 filas:")
print(data_problem4[:5])

# Separar las dimensiones
x2, y2 = data_problem4[:, 0], data_problem4[:, 1]

# Visualización combinada
plt.figure(figsize=(10, 8))
plt.scatter(x1, y1, alpha=0.6, c='blue', label='Problema 1 (0)', s=20)
plt.scatter(x2, y2, alpha=0.6, c='red', label='Problema 4 (1)', s=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Problema 4: Diagrama de Dispersión Combinado')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.show()

print(f"\nRangos de datos:")
print(f"Problema 1 - X: [{np.min(x1):.2f}, {np.max(x1):.2f}], Y: [{np.min(y1):.2f}, {np.max(y1):.2f}]")
print(f"Problema 4 - X: [{np.min(x2):.2f}, {np.max(x2):.2f}], Y: [{np.min(y2):.2f}, {np.max(y2):.2f}]")
```

```python
# ============================================================================
# PROBLEMA 5: Combinación de datos
# ============================================================================
print("\nPROBLEMA 5: Combinación de datos")
print("-" * 50)

# Método 1: usando np.vstack()
combined_data_vstack = np.vstack([data_problem1, data_problem4])
print(f"Forma usando np.vstack(): {combined_data_vstack.shape}")

# Método 2: usando np.concatenate()
combined_data_concat = np.concatenate([data_problem1, data_problem4], axis=0)
print(f"Forma usando np.concatenate(): {combined_data_concat.shape}")

# Verificar que ambos métodos dan el mismo resultado
print(f"¿Ambos métodos dan el mismo resultado? {np.array_equal(combined_data_vstack, combined_data_concat)}")

# Usar el resultado de vstack para continuar
combined_data = combined_data_vstack

print(f"\nPrimeras 5 filas del conjunto combinado:")
print(combined_data[:5])
print(f"Últimas 5 filas del conjunto combinado:")
print(combined_data[-5:])

print(f"\nEstadísticas del conjunto combinado:")
print(f"Media general: {np.mean(combined_data, axis=0)}")
print(f"Desviación estándar general: {np.std(combined_data, axis=0)}")
```

```python
# ============================================================================
# PROBLEMA 6: Etiquetado
# ============================================================================
print("\nPROBLEMA 6: Etiquetado")
print("-" * 50)

# Crear etiquetas: 0 para el primer conjunto, 1 para el segundo conjunto
labels_problem1 = np.zeros(500)  # 500 ceros
labels_problem4 = np.ones(500)   # 500 unos

# Combinar etiquetas
labels = np.concatenate([labels_problem1, labels_problem4])

print(f"Forma de las etiquetas: {labels.shape}")
print(f"Primeras 10 etiquetas: {labels[:10]}")
print(f"Últimas 10 etiquetas: {labels[-10:]}")
print(f"Número de etiquetas 0: {np.sum(labels == 0)}")
print(f"Número de etiquetas 1: {np.sum(labels == 1)}")

# Agregar columna de etiquetas usando np.column_stack()
final_dataset = np.column_stack([combined_data, labels])

print(f"\nForma del dataset final: {final_dataset.shape}")
print(f"Primeras 5 filas del dataset final:")
print(final_dataset[:5])
print(f"Últimas 5 filas del dataset final:")
print(final_dataset[-5:])

# Verificar estructura del dataset final
print(f"\nEstructura del dataset final:")
print(f"Columna 0 (X): rango [{np.min(final_dataset[:, 0]):.2f}, {np.max(final_dataset[:, 0]):.2f}]")
print(f"Columna 1 (Y): rango [{np.min(final_dataset[:, 1]):.2f}, {np.max(final_dataset[:, 1]):.2f}]")
print(f"Columna 2 (Etiquetas): valores únicos {np.unique(final_dataset[:, 2])}")
```

```python
# ============================================================================
# ANÁLISIS ADICIONAL Y VERIFICACIONES
# ============================================================================
print("\nANÁLISIS ADICIONAL")
print("-" * 50)

# Separar datos por etiqueta para análisis
data_class_0 = final_dataset[final_dataset[:, 2] == 0]
data_class_1 = final_dataset[final_dataset[:, 2] == 1]

print(f"Análisis por clases:")
print(f"Clase 0 - Cantidad: {len(data_class_0)}, Media: {np.mean(data_class_0[:, :2], axis=0)}")
print(f"Clase 1 - Cantidad: {len(data_class_1)}, Media: {np.mean(data_class_1[:, :2], axis=0)}")

# Visualización final del dataset completo
plt.figure(figsize=(12, 5))

# Subplot 1: Scatter plot por clases
plt.subplot(1, 2, 1)
plt.scatter(data_class_0[:, 0], data_class_0[:, 1], alpha=0.6, c='blue', label='Clase 0', s=20)
plt.scatter(data_class_1[:, 0], data_class_1[:, 1], alpha=0.6, c='red', label='Clase 1', s=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset Final: Dispersión por Clases')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 2: Distribución de etiquetas
plt.subplot(1, 2, 2)
unique_labels, counts = np.unique(final_dataset[:, 2], return_counts=True)
plt.bar(unique_labels, counts, color=['blue', 'red'], alpha=0.7)
plt.xlabel('Etiqueta')
plt.ylabel('Cantidad')
plt.title('Distribución de Etiquetas')
plt.xticks(unique_labels)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n✅ Dataset final creado exitosamente:")
print(f"   • Forma: {final_dataset.shape} (1000 muestras, 3 características)")
print(f"   • Características: X, Y, Etiqueta")
print(f"   • Clases balanceadas: {counts[0]} y {counts[1]} muestras")
print(f"   • Listo para uso en algoritmos de aprendizaje automático")
```

```python
# ============================================================================
# EXPORTAR DATOS (OPCIONAL)
# ============================================================================
print("\nEXPORTAR DATOS")
print("-" * 50)

# Guardar el dataset final en formato CSV
import pandas as pd

# Crear DataFrame para facilitar la exportación
df = pd.DataFrame(final_dataset, columns=['X', 'Y', 'Etiqueta'])
print("Primeras 10 filas del DataFrame:")
print(df.head(10))

# Guardar como CSV (descomenta la línea siguiente si quieres guardar)
df.to_csv('datos_ficticios.csv', index=False)
print("\n📊 Dataset listo para análisis de regresión lineal y otros algoritmos de ML")

# Resumen final
print(f"\n" + "="*60)
print("RESUMEN COMPLETADO")
print("="*60)
print(f"✅ Problema 1: 500 puntos generados con media {mean_1}")
print(f"✅ Problema 2: Scatter plot creado")
print(f"✅ Problema 3: Histogramas alineados creados")
print(f"✅ Problema 4: 500 puntos adicionales con media {mean_2}")
print(f"✅ Problema 5: Datos combinados usando np.vstack()")
print(f"✅ Problema 6: Etiquetas agregadas, dataset final (1000, 3)")
print(f"\n🎯 Dataset listo para aprendizaje automático!")
```
